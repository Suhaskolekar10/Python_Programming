# **1. String in Python**

---

A **string** is an **ordered sequence of characters** enclosed in **single `'` or double `"` quotes**.

```python
s1 = "Python"
s2 = 'Hello'
```

* Strings are **immutable** ‚Üí cannot be changed after creation.
* Strings are **sequences** ‚Üí you can access each character by **index**.

---

### üîπ String Indexing

Each character has an **index** starting from **0**:

```python
s = "Python"
print(s[0])  # P
print(s[5])  # n
```

* **Negative indexing** counts from the end:

```python
print(s[-1])  # n
print(s[-2])  # o
```

---

### üîπ String Slicing

Extract a **substring** using slicing:

```python
s[start:stop:step]
```

Examples:

```python
s = "PythonProgramming"

print(s[0:6])    # Python
print(s[6:])     # Programming
print(s[:6])     # Python
print(s[::2])    # PtoPormig (every 2nd char)
print(s[::-1])   # gnimmargorPnohtyP (reverse string)
```

---

### üîπ String Operations

#### 1. **Concatenation (`+`)**

```python
a = "Hello"
b = "World"
print(a + " " + b)  # Hello World
```

#### 2. **Repetition (`*`)**

```python
print("Hi! " * 3)  # Hi! Hi! Hi!
```

#### 3. **Membership**

```python
s = "Python"
print("Py" in s)      # True
print("Java" not in s) # True
```

#### 4. **Iteration**

```python
for ch in "Python":
    print(ch)
```

---

### üîπ String Methods

| Method                                  | Description                       | Example                                       |
| --------------------------------------- | --------------------------------- | --------------------------------------------- |
| `upper()`                               | Convert to uppercase              | `"abc".upper()` ‚Üí "ABC"                       |
| `lower()`                               | Convert to lowercase              | `"ABC".lower()` ‚Üí "abc"                       |
| `capitalize()`                          | First char uppercase              | `"hello".capitalize()` ‚Üí "Hello"              |
| `title()`                               | First char of each word uppercase | `"hello world".title()` ‚Üí "Hello World"       |
| `strip()`                               | Remove leading/trailing spaces    | `"  hi  ".strip()` ‚Üí "hi"                     |
| `lstrip()` / `rstrip()`                 | Left/Right strip                  | `"  hi  ".lstrip()` ‚Üí "hi  "                  |
| `replace(old, new)`                     | Replace substring                 | `"Python".replace("Py","Ja")` ‚Üí "Jathon"      |
| `split(sep)`                            | Split string into list            | `"a,b,c".split(",")` ‚Üí ['a','b','c']          |
| `join(iterable)`                        | Join iterable into string         | `" ".join(["Hello","World"])` ‚Üí "Hello World" |
| `find(sub)`                             | Index of substring                | `"Python".find("th")` ‚Üí 2                     |
| `count(sub)`                            | Count substring occurrences       | `"Python".count("o")` ‚Üí 1                     |
| `startswith(prefix)`                    | Check start                       | `"Python".startswith("Py")` ‚Üí True            |
| `endswith(suffix)`                      | Check end                         | `"Python".endswith("on")` ‚Üí True              |
| `isalpha()` / `isdigit()` / `isspace()` | Check type of string              | `"123".isdigit()` ‚Üí True                      |

---

### üîπ f-Strings (Formatted Strings)

Python 3.6+ feature ‚Äî **fast and readable string formatting**:

```python
name = "Suhas"
age = 22
print(f"My name is {name} and I am {age} years old.")
```

Other ways:

* `"Hello {}".format(name)`
* `"Hello %s" % name` (older style)

---

### üîπ Escape Characters

Special characters with `\`:

```python
print("Hello\nWorld")  # New line
print("It\'s Python")  # Single quote
print("Tab\tSpace")    # Tab space
```

---

### üîπ String Immutability

```python
s = "Python"
# s[0] = "J"  # ‚ùå Error
s = "J" + s[1:]  # ‚úÖ New string: "Jython"
```

---

### üîπ String Conversion

Convert other types to string:

```python
num = 10
s = str(num)  # '10'
```

Convert string to int/float:

```python
s = "123"
n = int(s)  # 123
f = float("3.14")  # 3.14
```

# **2. List in Python**
---

A **list** is an **ordered, mutable collection of items**.

* Can contain **different data types**: integers, strings, floats, even other lists.
* **Mutable** ‚Üí you can change, add, or remove elements.
* **Indexed** ‚Üí access elements by position (start from 0).

```python
my_list = [1, 2, 3, "Python", 4.5]
```

---

### üîπ List Creation

1. **Using square brackets**

```python
fruits = ["apple", "banana", "mango"]
```

2. **Using `list()` constructor**

```python
numbers = list([1, 2, 3, 4])
chars = list("Python")  # ['P','y','t','h','o','n']
```

---

### üîπ Accessing Elements

* **Indexing**

```python
fruits = ["apple", "banana", "mango"]
print(fruits[0])  # apple
print(fruits[-1]) # mango
```

* **Slicing**

```python
numbers = [1,2,3,4,5]
print(numbers[1:4])   # [2,3,4]
print(numbers[:3])    # [1,2,3]
print(numbers[::2])   # [1,3,5]
```

---

### üîπ List Operations

#### 1. Concatenation

```python
a = [1,2]
b = [3,4]
print(a + b)  # [1,2,3,4]
```

#### 2. Repetition

```python
print([0]*5)  # [0,0,0,0,0]
```

#### 3. Membership

```python
nums = [1,2,3]
print(2 in nums)      # True
print(5 not in nums)  # True
```

#### 4. Iteration

```python
for item in ["apple","banana"]:
    print(item)
```

---

### üîπ List Methods

| Method             | Description             | Example                             |
| ------------------ | ----------------------- | ----------------------------------- |
| `append(x)`        | Add element at end      | `[1,2].append(3)` ‚Üí `[1,2,3]`       |
| `insert(i, x)`     | Insert at index         | `[1,2].insert(1,5)` ‚Üí `[1,5,2]`     |
| `extend(iterable)` | Add multiple elements   | `[1,2].extend([3,4])` ‚Üí `[1,2,3,4]` |
| `remove(x)`        | Remove first occurrence | `[1,2,2].remove(2)` ‚Üí `[1,2]`       |
| `pop([i])`         | Remove & return element | `[1,2,3].pop()` ‚Üí 3                 |
| `index(x)`         | Return first index      | `[1,2,3].index(2)` ‚Üí 1              |
| `count(x)`         | Count occurrences       | `[1,2,2,3].count(2)` ‚Üí 2            |
| `sort()`           | Sort list ascending     | `[3,1,2].sort()` ‚Üí `[1,2,3]`        |
| `reverse()`        | Reverse list            | `[1,2,3].reverse()` ‚Üí `[3,2,1]`     |
| `copy()`           | Shallow copy            | `a.copy()`                          |
| `clear()`          | Remove all elements     | `[1,2].clear()` ‚Üí []                |

---

### üîπ Nested Lists

Lists can contain **other lists** ‚Äî useful for **matrices or grids**.

```python
matrix = [
    [1,2,3],
    [4,5,6],
    [7,8,9]
]

print(matrix[1][2])  # 6 (row 1, col 2)
```

---

### üîπ List Comprehensions

Pythonic way to **create or transform lists**:

```python
# Squares of numbers
nums = [1,2,3,4,5]
squares = [x**2 for x in nums]
print(squares)  # [1,4,9,16,25]

# Even numbers only
evens = [x for x in nums if x % 2 == 0]
print(evens)    # [2,4]
```

---

### üîπ Copying Lists

* **Shallow copy**

```python
a = [1,2,3]
b = a.copy()
```

* **Deep copy** (for nested lists)

```python
import copy
matrix2 = copy.deepcopy(matrix)
```

---

### üîπ Mutable vs Immutable Behavior

```python
a = [1,2,3]
b = a
b.append(4)
print(a)  # [1,2,3,4] ‚Üí same object, changed
```

‚úÖ Lists are **mutable**, unlike strings or tuples.

```python
nums = [10,20,30]
total = sum(nums)
print("Sum:", total)
```

1. **Difference between `append()` and `extend()`?**

   * `append()` ‚Üí adds a single element
   * `extend()` ‚Üí adds elements from an iterable

   * `mylist[::-1]` or `mylist.reverse()`

# **3. Tuple in Python**
---

A **tuple** is an **ordered, immutable collection** of items.

* **Ordered** ‚Üí elements have a defined sequence.
* **Immutable** ‚Üí cannot be changed after creation.
* Can contain **mixed data types**.
* **Indexed** ‚Üí access elements by position.

```python
my_tuple = (1, 2, 3, "Python", 4.5)
```

---

### üîπ Tuple Creation

1. **Using parentheses**

```python
t1 = (1, 2, 3)
```

2. **Without parentheses (tuple packing)**

```python
t2 = 4, 5, 6
```

3. **Single element tuple** ‚Üí needs a trailing comma

```python
t3 = (5,)   # Not (5)
```

4. **Using `tuple()` constructor**

```python
t4 = tuple([1,2,3])
t5 = tuple("Python")  # ('P','y','t','h','o','n')
```

---

### üîπ Accessing Tuple Elements

* **Indexing**

```python
t = (10,20,30)
print(t[0])   # 10
print(t[-1])  # 30
```

* **Slicing**

```python
print(t[:2])  # (10,20)
```

* **Iteration**

```python
for x in t:
    print(x)
```

---

### üîπ Tuple Operations

#### 1. Concatenation

```python
t1 = (1,2)
t2 = (3,4)
print(t1 + t2)  # (1,2,3,4)
```

#### 2. Repetition

```python
print(t1 * 2)  # (1,2,1,2)
```

#### 3. Membership

```python
print(2 in t1)      # True
print(5 not in t1)  # True
```

---

### üîπ Tuple Methods

| Method     | Description               | Example                |
| ---------- | ------------------------- | ---------------------- |
| `count(x)` | Count occurrences of x    | `(1,2,2).count(2)` ‚Üí 2 |
| `index(x)` | Index of first occurrence | `(1,2,3).index(2)` ‚Üí 1 |

‚úÖ Tuples have **fewer methods than lists** because they are immutable.

---

### üîπ Immutability of Tuples

```python
t = (1,2,3)
# t[0] = 5  # ‚ùå Error

# But mutable elements inside tuple can change
t2 = ([1,2], 3)
t2[0].append(3)  # ‚úÖ [1,2,3], tuple itself is immutable but list inside is mutable
```

---

### üîπ Tuple Packing and Unpacking

* **Packing**

```python
t = 10, 20, 30
```

* **Unpacking**

```python
a, b, c = t
print(a, b, c)  # 10 20 30
```

* Use `*` to capture remaining elements:

```python
a, *b = (1,2,3,4)
print(a)  # 1
print(b)  # [2,3,4]
```

---

### üîπ Tuple vs List

| Feature    | List    | Tuple      |
| ---------- | ------- | ---------- |
| Mutability | Mutable | Immutable  |
| Syntax     | `[]`    | `()`       |
| Methods    | Many    | Few        |
| Use case   | Dynamic | Fixed data |

---

### Example 5: Convert List to Tuple

```python
lst = [1,2,3]
t = tuple(lst)
print(t)  # (1,2,3)
```

# **4. Sets in Python**
---

A **set** is an **unordered collection of unique elements**.

* **Unordered** ‚Üí no indexing, elements have no fixed position
* **Unique** ‚Üí duplicates are automatically removed
* **Mutable** ‚Üí you can add or remove elements
* **Iterable** ‚Üí can loop over elements

```python
s = {1, 2, 3, 4}
```

---

### üîπ Set Creation

1. **Using curly braces `{}`**

```python
s1 = {1, 2, 3, 4}
```

2. **Using `set()` constructor**

```python
s2 = set([1, 2, 2, 3])  # {1, 2, 3}
s3 = set("Python")       # {'P','y','t','h','o','n'}
```

‚úÖ Note: **Empty set** must be created using `set()`, not `{}` (that creates an empty dict).

```python
empty_set = set()
```

---

### üîπ Accessing Elements

* **Unordered** ‚Üí no indexing
* **Iteration**:

```python
s = {1,2,3}
for elem in s:
    print(elem)
```

---

### üîπ Set Operations (Mathematical)

| Operation      | Symbol / Method                | Example                   |        |                  |
| -------------- | ------------------------------ | ------------------------- | ------ | ---------------- |
| Union          | `                              | `/`union()`               | `{1,2} | {2,3}`‚Üí`{1,2,3}` |
| Intersection   | `&` / `intersection()`         | `{1,2} & {2,3}` ‚Üí `{2}`   |        |                  |
| Difference     | `-` / `difference()`           | `{1,2} - {2,3}` ‚Üí `{1}`   |        |                  |
| Symmetric Diff | `^` / `symmetric_difference()` | `{1,2} ^ {2,3}` ‚Üí `{1,3}` |        |                  |

---

### üîπ Adding and Removing Elements

| Method             | Description                            | Example           |
| ------------------ | -------------------------------------- | ----------------- |
| `add(x)`           | Add a single element                   | `s.add(5)`        |
| `update(iterable)` | Add multiple elements                  | `s.update([6,7])` |
| `remove(x)`        | Remove element (error if missing)      | `s.remove(2)`     |
| `discard(x)`       | Remove element (no error if missing)   | `s.discard(10)`   |
| `pop()`            | Remove and return an arbitrary element | `s.pop()`         |
| `clear()`          | Remove all elements                    | `s.clear()`       |

---

### üîπ Set Methods ‚Äî Quick Summary

| Method                   | Description                                        |
| ------------------------ | -------------------------------------------------- |
| `union()`                | Combine sets (no duplicates)                       |

| `difference()`           | Elements in first set but not second               |

| `issubset()`             | Checks if all elements are in another set          |
| `issuperset()`           | Checks if set contains all elements of another set |

# **5. Dictionary in Python**
---

A **dictionary** is an **unordered, mutable collection** of **key-value pairs**.

* Each **key** is **unique** and **immutable** (string, number, or tuple).
* Each **value** can be **any data type**.
* Think of it as a **real-world dictionary** ‚Äî word = key, definition = value.

```python
person = {
    "name": "Suhas",
    "age": 22,
    "skills": ["Python", "Data Analysis"]
}
```

---

### üîπ Dictionary Characteristics

| Feature                   | Description                             |
| ------------------------- | --------------------------------------- |
| **Ordered (Python 3.7+)** | Preserves insertion order               |
| **Mutable**               | Can add, modify, or remove items        |
| **Key uniqueness**        | Duplicate keys overwrite previous ones  |
| **Key immutability**      | Keys must be immutable (no lists, sets) |

---

### üîπ Creating Dictionaries

1. **Using `{}`**

```python
d = {"name": "Suhas", "age": 22}
```

2. **Using `dict()` constructor**

```python
d = dict(name="Suhas", age=22)
```

3. **From list of tuples**

```python
pairs = [("name","Suhas"),("age",22)]
d = dict(pairs)
```

4. **Empty dictionary**

```python
d = {}
```

---

### üîπ Accessing and Modifying Values

#### Access value by key

```python
print(d["name"])  # Suhas
```

#### Using `.get()` (avoids KeyError)

```python
print(d.get("name"))      # Suhas
print(d.get("city", "NA"))  # Returns default value
```

#### Add or update key-value

```python
d["city"] = "Hyderabad"
d["age"] = 23  # update existing
```

#### Remove elements

```python
d.pop("age")        # remove key and return value
d.popitem()         # remove last inserted item (Python 3.7+)
del d["city"]       # delete key
d.clear()           # remove all
```

---

### üîπ Dictionary Methods (Quick Table)

| Method              | Description                        |
| ------------------- | ---------------------------------- |
| `keys()`            | Returns all keys                   |
| `values()`          | Returns all values                 |
| `items()`           | Returns list of (key, value) pairs |
| `get(key, default)` | Returns value if key exists        |
| `update(d2)`        | Merges another dictionary          |
| `pop(key)`          | Removes specific key               |
| `popitem()`         | Removes last inserted key          |
| `clear()`           | Empties dictionary                 |
| `copy()`            | Returns shallow copy               |

---

### üîπ Iterating Over a Dictionary

```python
person = {"name": "Suhas", "age": 22, "city": "Hyd"}

for key in person:
    print(key, person[key])

for key, value in person.items():
    print(f"{key} ‚Üí {value}")
```

---

### üîπ Dictionary Comprehensions

Like list comprehensions but for key-value pairs üî•

```python
squares = {x: x**2 for x in range(5)}
print(squares)  # {0:0, 1:1, 2:4, 3:9, 4:16}
```

Conditional comprehension:

```python
even_squares = {x: x**2 for x in range(10) if x%2==0}
```

---

### üîπ Nested Dictionaries

```python
student = {
    "name": "Suhas",
    "subjects": {
        "math": 95,
        "science": 90
    }
}

print(student["subjects"]["math"])  # 95
```

---

### üîπ Dictionary Unpacking

```python
a = {"x": 1, "y": 2}
b = {"y": 3, "z": 4}

merged = {**a, **b}
print(merged)  # {'x':1, 'y':3, 'z':4}
```

## üí° **Real-Life Implementation Examples**

### Example 1: Word Frequency Counter

```python
sentence = "python is fun and python is easy"
freq = {}
for word in sentence.split():
    freq[word] = freq.get(word, 0) + 1
print(freq)
# {'python': 2, 'is': 2, 'fun': 1, 'and': 1, 'easy': 1}
```

### Example 2: Student Marks Tracker

```python
students = {
    "Alice": {"math": 90, "science": 85},
    "Bob": {"math": 80, "science": 88}
}
print(students["Alice"]["science"])  # 85
```

### Example 3: Convert Two Lists to Dictionary

```python
keys = ["name","age","city"]
values = ["Suhas",22,"Hyderabad"]
d = dict(zip(keys, values))
print(d)
```

### Example 4: JSON-like API Data

```python
user = {"id":101, "name":"Suhas", "verified":True}
print(user["name"])
```

### Example 5: Merge Dictionaries

```python
a = {"x":1, "y":2}
b = {"y":3, "z":4}
merged = a | b
print(merged)
```
