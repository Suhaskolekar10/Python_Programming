* **Python 2.0 (2000)** – List comprehensions, garbage collection.
* **Python 3.0 (2008)** – Major redesign; not backward compatible with Python 2.
* **Python 2.x officially ended support in 2020** → Python 3 is the standard now.

### **Philosophy**

* Emphasizes **readability**, **simplicity**, and **developer productivity**.
* Inspired by ABC language; uses clear English-like syntax.
* "There should be one—and preferably only one—obvious way to do it" (The Zen of Python).

# **5. Operators in Python**
---

Operators are special symbols or keywords used to perform operations on variables and values.

Python has the following types of operators:

1. **Arithmetic Operators**
2. **Comparison (Relational) Operators**
3. **Logical Operators**
4. **Assignment Operators**
5. **Bitwise Operators**
6. **Membership Operators**
7. **Identity Operators**

---

# ⭐ **1. Arithmetic Operators**

Used for mathematical operations.

| Operator | Meaning             | Example  |
| -------- | ------------------- | -------- |
| `+`      | Addition            | `a + b`  |
| `-`      | Subtraction         | `a - b`  |
| `*`      | Multiplication      | `a * b`  |
| `/`      | Division (float)    | `a / b`  |
| `//`     | Floor division      | `a // b` |
| `%`      | Modulus (remainder) | `a % b`  |
| `**`     | Exponentiation      | `a ** b` |

### Example:

```python
a = 10
b = 3
print(a / b)   # 3.333
print(a // b)  # 3
print(a % b)   # 1
print(a ** b)  # 1000
```

---

# ⭐ **2. Comparison Operators**

Used to compare two values. Output is boolean (`True`/`False`).

| Operator | Meaning                  |
| -------- | ------------------------ |
| `==`     | Equal to                 |
| `!=`     | Not equal to             |
| `>`      | Greater than             |
| `<`      | Less than                |
| `>=`     | Greater than or equal to |
| `<=`     | Less than or equal to    |

### Example:

```python
5 > 3   # True
5 == 5  # True
7 != 7  # False
```

---

# ⭐ **3. Logical Operators**

Used to combine conditional statements.

| Operator | Meaning                                | Example            |
| -------- | -------------------------------------- | ------------------ |
| `and`    | True if both conditions are True       | `a > 5 and b < 10` |
| `or`     | True if at least one condition is True | `a > 5 or b < 10`  |
| `not`    | Reverses True ↔ False                  | `not(a > 5)`       |

### Example:

```python
x = 5
print(x > 3 and x < 10)  # True
print(x > 3 or x > 10)   # True
print(not(x > 3))        # False
```

---

# ⭐ **4. Assignment Operators**

Used to assign values to variables.

| Operator | Meaning                 | Example   |
| -------- | ----------------------- | --------- |
| `=`      | Assign                  | `x = 5`   |
| `+=`     | Add and assign          | `x += 3`  |
| `-=`     | Subtract and assign     | `x -= 3`  |
| `*=`     | Multiply and assign     | `x *= 3`  |
| `/=`     | Divide and assign       | `x /= 3`  |
| `//=`    | Floor divide and assign | `x //= 3` |
| `%=`     | Modulus and assign      | `x %= 3`  |
| `**=`    | Exponent assign         | `x **= 3` |

### Example:

```python
x = 10
x += 5   # 15
x *= 2   # 30
```

---

# ⭐ **5. Bitwise Operators**

Used to work at the **binary level**.

| Operator | Meaning     |    |
| -------- | ----------- | -- |
| `&`      | AND         |    |
| `        | `           | OR |
| `^`      | XOR         |    |
| `~`      | NOT         |    |
| `<<`     | Left shift  |    |
| `>>`     | Right shift |    |

### Example:

```python
a = 5   # 0101
b = 3   # 0011
print(a & b)  # 1 (0001)
print(a | b)  # 7 (0111)
```

---

# ⭐ **6. Membership Operators**

Check if a value exists in a sequence (list, string, tuple, set, dict).

| Operator | Meaning                      |
| -------- | ---------------------------- |
| `in`     | True if value exists         |
| `not in` | True if value does NOT exist |

### Example:

```python
fruits = ["apple", "banana"]
print("apple" in fruits)      # True
print("mango" not in fruits)  # True
```

---

# ⭐ **7. Identity Operators**

Used to compare **memory location** of objects.

| Operator | Meaning                                   |
| -------- | ----------------------------------------- |
| `is`     | True if both objects refer to same memory |
| `is not` | True if memory location is different      |

### Example:

```python
a = [1, 2]
b = [1, 2]
print(a == b)   # True (values equal)
print(a is b)   # False (different objects)
```

---

# ⭐ **8. Operator Precedence**

Priority of operations:

1. `()`
2. `**`
3. `*, /, //, %`
4. `+, -`
5. `>, <, ==, != …`
6. `not`
7. `and`
8. `or`

### Example:

```python
print(5 + 2 * 3)     # 11
print((5 + 2) * 3)   # 21
```

---

# ✔ **Quick Summary**

* Arithmetic → math operations
* Comparison → boolean result
* Logical → combine conditions
* Assignment → update variable value
* Bitwise → binary operations
* Membership → check existence
* Identity → compare memory
* Precedence → defines execution order

# **6. Conditional Statements in Python**
----

Conditional statements allow a program to **make decisions** based on certain conditions.

Python uses:

1. `if`
2. `elif`
3. `else`

---

# ⭐ **1. The `if` Statement**

Used to execute a block of code **only if** a condition is True.

### Syntax:

```python
if condition:
    statement(s)
```

### Example:

```python
age = 18
if age >= 18:
    print("You are eligible to vote")
```

---

# ⭐ **2. The `else` Statement**

Runs when the **if condition is False**.

### Syntax:

```python
if condition:
    statements
else:
    statements
```

### Example:

```python
x = 5
if x > 10:
    print("Greater")
else:
    print("Not greater")
```

---

# ⭐ **3. The `elif` Statement**

Used when you need **multiple conditions**.

### Syntax:

```python
if condition1:
    statements
elif condition2:
    statements
elif condition3:
    statements
else:
    statements
```

### Example:

```python
marks = 75

if marks >= 90:
    print("Grade A")
elif marks >= 75:
    print("Grade B")
elif marks >= 50:
    print("Grade C")
else:
    print("Fail")
```

---

# ⭐ **4. Important Rules**

### ✔ Conditions must be Boolean expressions

They evaluate to `True` or `False`.

Example:

```python
if x == 10:
    ...
```

### ✔ Each condition ends with a colon `:`

Python uses indentation for blocks.

### ✔ Only one block executes

Once a matching condition is found, others are ignored.

---

# ⭐ **5. Nested `if` Statements**

An `if` inside another `if`.

### Example:

```python
age = 20
license = True

if age >= 18:
    if license:
        print("You can drive")
    else:
        print("Get a license first")
else:
    print("You are too young")
```

---

# ⭐ **6. Using Logical Operators in Conditions**

### `and`, `or`, `not` with `if`

Example:

```python
age = 25
citizen = True

if age >= 18 and citizen:
    print("Eligible to vote")
```

---

# ⭐ **7. Short-Hand `if` (Single Line)**

```python
x = 5
if x > 0: print("Positive")
```

---

# ⭐ **8. Short-Hand `if-else` (Ternary Operator)**

```python
a = 10
b = 20

result = "A is bigger" if a > b else "B is bigger"
print(result)
```

---

# ⭐ **9. Multiple Conditions in One Line**

```python
x = 15

if 10 < x < 20:
    print("In range")
```

---

# ⭐ **10. Real-Life Example Program**

```python
temp = int(input("Enter temperature: "))

if temp > 30:
    print("Hot")
elif temp > 20:
    print("Warm")
elif temp > 10:
    print("Cool")
else:
    print("Cold")
```

---

# ✔ **Quick Summary (Short Notes)**

* **if** → checks the first condition
* **elif** → checks additional conditions
* **else** → runs when all conditions are false
* Blocks are defined using **indentation**
* Conditions must be **True/False**
* Use logical operators for complex decisions

# **7. Loops in Python**
---

Loops are used to repeat a block of code multiple times.

Python mainly provides two types of loops:

1. **for loop**
2. **while loop**

Additionally:

* `break`
* `continue`
* `pass`

---

# ⭐ **1. `for` Loop**

Used to iterate over sequences like:

* List
* String
* Tuple
* Set
* Dictionary
* range()

### **Syntax:**

```python
for variable in sequence:
    statements
```

---

### ✔ Example 1: Loop through a list

```python
fruits = ["apple", "banana", "mango"]
for fruit in fruits:
    print(fruit)
```

---

### ✔ Example 2: Loop through a string

```python
for ch in "Python":
    print(ch)
```

---

### ✔ Example 3: Loop using `range()`

`range(start, end, step)`
(end is excluded)

```python
for i in range(1, 6):
    print(i)
```

---

### ✔ Example 4: Using step

```python
for i in range(0, 10, 2):
    print(i)
```

---

# ⭐ **2. `while` Loop**

Repeats as long as a condition is True.

### **Syntax:**

```python
while condition:
    statements
```

---

### ✔ Example:

```python
i = 1
while i <= 5:
    print(i)
    i += 1
```

---

### ⚠ Infinite Loop Example

```python
while True:
    print("Runs forever")
```

(Use Ctrl+C to stop in terminal)

---

# ⭐ **3. Loop Control Statements**

## ✔ **`break`**

Stops loop immediately.

```python
for i in range(10):
    if i == 5:
        break
    print(i)
```

---

## ✔ **`continue`**

Skips the current iteration and continues the loop.

```python
for i in range(6):
    if i == 3:
        continue
    print(i)
```

---

## ✔ **`pass`**

Does nothing → used as a placeholder.

```python
for i in range(5):
    pass  # later logic will be added
```

---

# ⭐ **4. `else` with Loops**

`else` executes when loop completes normally
(closed without `break`).

### ✔ Example (for loop):

```python
for i in range(5):
    print(i)
else:
    print("Loop finished")
```

### ✔ Example (break stops else):

```python
for i in range(5):
    if i == 2:
        break
else:
    print("Won't run")
```

---

# ⭐ **5. Nested Loops**

A loop inside another loop.

```python
for i in range(3):
    for j in range(3):
        print(i, j)
```

---

# ⭐ **6. Looping through dictionaries**

```python
student = {"name": "Suhas", "age": 22}

for key in student:
    print(key, student[key])
```

Or:

```python
for k, v in student.items():
    print(k, v)
```

# **8. Functions in Python**
---

A **function** is a block of reusable code that performs a specific task.
Functions help to write clean, modular, and reusable programs.

---

# ⭐ 1. **Types of Functions in Python**

Python has two types:

1. **Built-in functions**
   Examples: `print()`, `len()`, `input()`, `type()`, `max()`, etc.

2. **User-defined functions**
   Functions created by the programmer using `def`.

---

# ⭐ 2. **Defining a Function**

### **Syntax:**

```python
def function_name(parameters):
    statements
```

### **Example:**

```python
def greet():
    print("Hello, Python!")
```

### Call the function:

```python
greet()
```

---

# ⭐ 3. **Function with Parameters**

### Example:

```python
def add(a, b):
    print(a + b)
```

### Calling:

```python
add(5, 7)
```

---

# ⭐ 4. **Function with Return Value**

`return` sends value back to caller.

```python
def multiply(x, y):
    return x * y
```

```python
result = multiply(4, 5)
print(result)
```

Without `return`, a function returns `None`.

---

# ⭐ 5. **Types of Arguments**

Python supports:

### ✔ **1. Positional Arguments**

Order matters.

```python
def student(name, age):
    print(name, age)

student("Suhas", 21)
```

---

### ✔ **2. Keyword Arguments**

Order doesn't matter.

```python
student(age=21, name="Suhas")
```

---

### ✔ **3. Default Arguments**

Used when a value is not provided.

```python
def greet(name="User"):
    print("Hello", name)

greet()
greet("Suhas")
```

---

### ✔ **4. Variable-Length Arguments**

### **a) *args** → for multiple positional arguments

Stored as a **tuple**

```python
def total(*numbers):
    print(sum(numbers))

total(10, 20, 30)
```

---

### **b) **kwargs** → multiple keyword arguments

Stored as a **dictionary**

```python
def info(**data):
    print(data)

info(name="Suhas", age=21, city="Pune")
```

---

# ⭐ 6. **Return Multiple Values**

Python allows returning multiple values using tuple packing.

```python
def calc(a, b):
    return a+b, a-b, a*b

x, y, z = calc(10, 5)
print(x, y, z)
```

---

# ⭐ 7. **Local and Global Variables**

### ✔ Local Variable

Declared inside a function → available only there.

```python
def func():
    x = 10  # local
    print(x)
```

---

### ✔ Global Variable

Declared outside functions → accessible everywhere.

```python
x = 100

def show():
    print(x)

show()
```

---

### **Using global keyword**

```python
x = 5

def change():
    global x
    x = 20

change()
print(x)
```

---

# ⭐ 8. **Anonymous Functions (Lambda)**

Functions without a name.

### Syntax:

```python
lambda args: expression
```

### Example:

```python
square = lambda n: n * n
print(square(6))
```

Multiple arguments:

```python
add = lambda a, b: a + b
print(add(3, 4))
```

---

# ⭐ 9. **Docstrings (Function Documentation)**

Used to describe a function.

```python
def add(a, b):
    """This function returns the sum of two numbers."""
    return a + b
```

Access docstring:

```python
print(add.__doc__)
```

---

# ⭐ 10. **Nested Functions**

A function inside another function.

```python
def outer():
    def inner():
        print("Inside inner")
    inner()

outer()
```

---

# ⭐ 11. **Recursion**

A function calling itself.

```python
def fact(n):
    if n == 1:
        return 1
    return n * fact(n-1)

print(fact(5))
```
