# 📌 1. **math module** (Built-in)

Provides mathematical functions like square root, trigonometry, constants, rounding, etc.

### ✨ Common Functions

```python
import math
```

### 👇 Basic Math

```python
math.sqrt(25)         # 5.0
math.pow(2, 3)        # 8.0
math.factorial(5)     # 120
math.ceil(4.1)        # 5
math.floor(4.9)       # 4
math.fabs(-10)        # 10.0
```

### 👇 Trigonometry

```python
math.sin(math.pi/2)   # 1.0
math.cos(0)           # 1.0
math.tan(math.pi/4)   # 1.0
```

### 👇 Constants

```python
math.pi               # 3.14159...
math.e                # 2.71828...
```

# 📌 2. **random module** (Built-in)

Used for generating random numbers, picking random items, shuffling, etc.

```python
import random
```

### 👇 Random Numbers

```python
random.random()          # float between 0 and 1
random.randint(1, 10)    # integer between 1 to 10
random.uniform(1, 5)     # float between range
```

### 👇 Pick from a list

```python
random.choice(['red', 'blue', 'green'])
```

### 👇 Shuffle a list

```python
items = [1, 2, 3, 4]
random.shuffle(items)
```

### 👇 Pick multiple random items

```python
random.sample([1,2,3,4,5], 2)    # picks 2 items
```

# 📌 3. **datetime module** (Built-in)

Used to work with dates, times, differences between dates, formatting, etc.

```python
from datetime import datetime, date, time, timedelta
```

### 👇 Current date & time

```python
now = datetime.now()
print(now)
```

### 👇 Create custom date/time

```python
d = date(2025, 1, 1)
t = time(12, 30, 0)
```

### 👇 Formatting dates

```python
now.strftime("%Y-%m-%d %H:%M:%S")
```

Common format codes:

| Code | Meaning    | Example |
| ---- | ---------- | ------- |
| `%Y` | Full year  | 2025    |
| `%m` | Month      | 09      |
| `%d` | Day        | 23      |
| `%H` | Hour (24h) | 14      |
| `%M` | Minute     | 30      |
| `%S` | Second     | 45      |

### 👇 Parse string → date

```python
dt = datetime.strptime("2025-11-05", "%Y-%m-%d")
```

### 👇 Date arithmetic

```python
next_week = datetime.now() + timedelta(days=7)
```

## ⭐ Working with DataFrames

### 👇 Create a DataFrame

```python
data = {
    "Name": ["A", "B", "C"],
    "Age": [25, 30, 22]
}

df = pd.DataFrame(data)
print(df)
```

---

### 👇 Read CSV File

```python
df = pd.read_csv("data.csv")
```

---

### 👇 Show basic info

```python
df.head()        # first 5 rows
df.tail()        # last 5 rows
df.info()        # summary
df.describe()    # statistics
```

---

### 👇 Select columns

```python
df["Age"]
df[["Name", "Age"]]
```

---

### 👇 Filter rows

```python
df[df["Age"] > 25]
```

---

### 👇 Add new column

```python
df["NewCol"] = df["Age"] + 5
```

---

### 👇 Save data

```python
df.to_csv("output.csv", index=False)
```
