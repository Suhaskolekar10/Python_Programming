# **1. Strings in Python**
----

A **string** is a sequence of characters enclosed in quotes.

Example:

```python
name = "Python"
```

Strings are:

* **Immutable** (cannot be changed)
* **Ordered / Indexed**
* **Iterable**
* **Support slicing**
* **Support many built-in methods**

---

# ⭐ 1. **Creating Strings**

Python allows three types of quotes:

```python
s1 = 'Hello'
s2 = "Hello"
s3 = '''Hello'''
```

Triple quotes (`'''` or `"""`) allow **multi-line strings**:

```python
text = """This is
a multi-line string."""
```

---

# ⭐ 2. **Accessing Characters**

### ✔ Indexing

Python indexing starts from **0**.

```python
s = "Python"
print(s[0])   # P
print(s[5])   # n
```

### ✔ Negative Indexing

```python
s = "Python"
print(s[-1])  # n
print(s[-6])  # P
```

---

# ⭐ 3. **Slicing Strings**

Syntax:
`string[start:end:step]`

```python
s = "Python"
print(s[1:4])      # yth
print(s[:3])       # Pyt
print(s[3:])       # hon
print(s[::-1])     # reverse string → nohtyP
```

---

# ⭐ 4. **String Properties**

### ✔ **Strings are immutable**

You *cannot* modify individual characters.

❌ Invalid:

```python
s = "Hello"
s[0] = 'Y'  # Error
```

✔ Valid (create new string):

```python
s = "Hello"
s = "Y" + s[1:]
```

---

# ⭐ 5. **String Concatenation and Repetition**

### ✔ Concatenation:

```python
print("Hello" + "World")
```

### ✔ Repetition:

```python
print("Hi" * 3)    # HiHiHi
```

---

# ⭐ 6. **String Methods**

Python provides many built-in methods for string operations.

---

## 🔵 **1) Case Methods**

```python
text = "python"
text.upper()      # PYTHON
text.lower()      # python
text.title()      # Python
text.capitalize() # Python
text.swapcase()   # PYTHON → python
```

---

## 🔵 **2) Checking Methods (Return True/False)**

```python
s.isalpha()   # alphabets only?
s.isdigit()   # digits only?
s.isalnum()   # alphanumeric?
s.islower()
s.isupper()
s.isspace()
```

---

## 🔵 **3) Searching & Counting**

```python
s = "hello world"
s.find("o")         # returns index or -1
s.rfind("o")        # search from right
s.count("l")        # count occurrences
```

---

## 🔵 **4) Replace**

```python
s.replace("hello", "hi")
```

---

## 🔵 **5) Strip (Remove Spaces)**

```python
s = "  python  "
s.strip()      # both sides
s.lstrip()     # left
s.rstrip()     # right
```

---

## 🔵 **6) Split & Join**

### ✔ split()

Convert string → list:

```python
"hello world".split()         # ['hello', 'world']
"apple,banana".split(",")     # ['apple', 'banana']
```

### ✔ join()

List → string:

```python
words = ['hello', 'world']
" ".join(words)               # hello world
```

---

## 🔵 **7) Startswith / Endswith**

```python
s.startswith("Py")
s.endswith("on")
```

---

# ⭐ 7. **String Formatting**

### ✔ f-string (modern & recommended)

```python
name = "Suhas"
age = 21
print(f"My name is {name} and I am {age} years old.")
```

---

### ✔ format() method

```python
"Hello {}, you are {} years old".format("Suhas", 21)
```

---

### ✔ % formatting (old)

```python
"%s is %d years old" % ("Suhas", 21)
```

---

# ⭐ 8. **Escape Characters**

Used to represent special characters.

| Escape | Meaning      |
| ------ | ------------ |
| `\n`   | New line     |
| `\t`   | Tab          |
| `\"`   | Double quote |
| `\'`   | Single quote |
| `\\`   | Backslash    |

Example:

```python
print("Hello\nWorld")
```

---

# ⭐ 9. **Iterating Through a String**

```python
for ch in "Python":
    print(ch)
```

---

# ⭐ 10. **Important String Operations**

### ✔ Reverse a string

```python
s[::-1]
```

### ✔ Check palindrome

```python
s == s[::-1]
```

### ✔ Count vowels

```python
sum(ch in 'aeiouAEIOU' for ch in s)
```

---

# ⭐ 11. **String Comparison**

Python compares lexicographically (dictionary order).

```python
"apple" < "banana"    # True
```

Case-sensitive:

```python
"A" < "a"     # True (because ASCII value smaller)
```

---

# ⭐ 12. **Multiline Strings**

```python
msg = """This is
a multiline
message."""
```

---

# ⭐ 13. **Raw Strings**

Ignores escape sequences (important in file paths & regex).

```python
r"C:\new\folder"
```

---

# ⭐ 14. **Short Summary**

* Strings are **immutable sequences** of characters.
* Support **indexing, slicing, iterating, concatenation, repetition**.
* Provide over **40+ built-in methods**.
* Used heavily in:

  * text processing
  * data cleaning
  * file handling
  * web & API data
  * natural language processing

# **2. List in Python**
----

A **list** is an ordered, mutable (changeable) collection of items.

Example:

```python
my_list = [10, 20, 30]
```

---

# ⭐ 1. **Characteristics of Lists**

✔ **Ordered** → maintains the order of insertion

✔ **Mutable** → can modify elements (add, delete, update)

✔ **Heterogeneous** → can store different data types

✔ **Index-based** → supports positive & negative indexing

✔ **Allows duplicates**

✔ **Dynamic** → size can grow/shrink

---

# ⭐ 2. **Creating Lists**

```python
l1 = [1, 2, 3]
l2 = ["apple", "banana"]
l3 = [1, "hello", 3.5, True]
l4 = []            # Empty list
l5 = list((1, 2, 3))  # Using list() constructor
```

---

# ⭐ 3. **Accessing List Elements**

### ✔ Indexing

```python
l = [10, 20, 30]
print(l[0])    # 10
print(l[-1])   # 30
```

### ✔ Slicing

```python
l = [10, 20, 30, 40, 50]
print(l[1:4])      # [20, 30, 40]
print(l[::-1])     # Reverse list
```

---

# ⭐ 4. **Updating List Elements**

Lists are mutable:

```python
l = [10, 20, 30]
l[1] = 200
```

---

# ⭐ 5. **List Methods (Very Important)**

Here are all major list methods with examples.

---

## 🔵 **1. append() – Add at end**

```python
l = [1, 2, 3]
l.append(4)
```

---

## 🔵 **2. insert() – Add at specific index**

```python
l.insert(1, "hello")
```

---

## 🔵 **3. extend() – Add multiple elements**

```python
l = [1, 2]
l.extend([3, 4, 5])
```

---

## 🔵 **4. remove() – Remove first occurrence**

```python
l.remove(2)
```

---

## 🔵 **5. pop() – Remove by index (returns removed item)**

```python
l.pop(1)   # removes element at index 1
l.pop()    # removes last element
```

---

## 🔵 **6. clear() – Empty the list**

```python
l.clear()
```

---

## 🔵 **7. index() – Find index of element**

```python
l.index(30)
```

---

## 🔵 **8. count() – Count occurrences**

```python
l.count(5)
```

---

## 🔵 **9. sort() – Sort list**

```python
l.sort()           # Ascending
l.sort(reverse=True)   # Descending
```

---

## 🔵 **10. reverse()**

```python
l.reverse()
```

---

# ⭐ 6. **List Operations**

### ✔ Concatenation

```python
l3 = l1 + l2
```

### ✔ Repetition

```python
l = [1, 2] * 3
```

### ✔ Membership test

```python
2 in [1,2,3]    # True
```

---

# ⭐ 7. **Iterating Over Lists**

### For loop:

```python
for x in [10, 20, 30]:
    print(x)
```

### While loop:

```python
i = 0
while i < len(l):
    print(l[i])
    i += 1
```

---

# ⭐ 8. **List Comprehension (Most Important)**

A short way to create lists.

### ✔ Basic:

```python
squares = [x*x for x in range(5)]
```

### ✔ With condition:

```python
evens = [x for x in range(10) if x%2==0]
```

### ✔ Nested:

```python
pairs = [(x, y) for x in range(3) for y in range(3)]
```

---

# ⭐ 9. **Nested Lists**

Lists inside lists.

```python
matrix = [
    [1,2,3],
    [4,5,6],
    [7,8,9]
]
print(matrix[1][2])   # 6
```

---

# ⭐ 10. **Copying Lists**

### ✔ Shallow Copy:

```python
a = [1,2,3]
b = a.copy()
```

OR:

```python
b = a[:]
```

### ✔ Deep Copy (for nested lists):

```python
import copy
b = copy.deepcopy(a)
```

---

# ⭐ 11. **Deleting Elements**

```python
del l[1]     # delete index
del l[1:3]   # delete range
del l        # delete entire list
```

---

# ⭐ 12. **List vs Tuple (Quick Difference)**

| Feature    | List         | Tuple      |
| ---------- | ------------ | ---------- |
| Mutability | Mutable      | Immutable  |
| Methods    | Many         | Few        |
| Speed      | Slower       | Faster     |
| Use Case   | Dynamic data | Fixed data |

---

# ⭐ 13. **Common Real-World Uses of Lists**

✔ Store records
✔ Dynamic data storage
✔ Matrices
✔ Queue & stack implementation
✔ Data manipulation in ML/NLP
✔ Processing user input

---

# ⭐ 14. **Important Examples**

### ✔ Find largest & smallest:

```python
max(l)
min(l)
```

### ✔ Sum of list:

```python
sum(l)
```

### ✔ Remove duplicates:

```python
l = list(set(l))
```

---

# ⭐ 15. **Short Notes (For Revision)**

* Lists are dynamic, ordered, and mutable
* Supports slicing, indexing, updating
* Main methods: append(), extend(), insert(), pop(), remove(), sort(), reverse()
* Supports list comprehension

# **3. Tuples in Python**
---

A **tuple** is an ordered, immutable collection of elements.

Example:

```python
t = (1, 2, 3)
```

---

# ⭐ 1. **What is a Tuple?**

A tuple is:

* **Ordered** → elements have fixed positions
* **Immutable** → cannot be changed after creation
* **Allows duplicates**
* **Allows heterogeneous data** (int, float, string, list, etc.)
* **Faster than lists**
* **Hashable** (can be used as dictionary keys if elements are immutable)

---

# ⭐ 2. **Creating a Tuple**

### ✔ Using parentheses:

```python
t = (1, 2, 3)
```

### ✔ Without parentheses (tuple packing):

```python
t = 1, 2, 3
```

### ✔ Single element tuple (comma needed):

```python
t = (10,)   # Correct
```

❌ Wrong:

```python
t = (10)     # This is an int
```

### ✔ Empty tuple:

```python
t = ()
```

---

# ⭐ 3. **Tuple Indexing**

```python
t = ("a", "b", "c")
print(t[0])   # a
print(t[-1])  # c
```

---

# ⭐ 4. **Tuple Slicing**

```python
t = (10, 20, 30, 40, 50)
print(t[1:4])     # (20, 30, 40)
print(t[::-1])    # reversed tuple
```

---

# ⭐ 5. **Tuple Immutability**

Tuples **cannot** be changed after creation.

❌ Not allowed:

```python
t = (1, 2, 3)
t[0] = 5   # Error
```

✔ But you can reassign:

```python
t = (1, 2, 3)
t = (4, 5, 6)
```

---

# ⭐ 6. **Accessing Elements with Loop**

```python
for x in (10, 20, 30):
    print(x)
```

---

# ⭐ 7. **Tuple Methods**

Tuples have only **2** built-in methods:

### ✔ count()

```python
t = (1, 2, 2, 3)
t.count(2)    # 2
```

### ✔ index()

```python
t = (10, 20, 30)
t.index(20)   # 1
```

---

# ⭐ 8. **Tuple Operations**

### ✔ Concatenation

```python
t1 = (1, 2)
t2 = (3, 4)
t3 = t1 + t2
```

### ✔ Repetition

```python
t = (1, 2)
print(t * 3)   # (1,2,1,2,1,2)
```

### ✔ Membership Test

```python
3 in (1, 2, 3)       # True
```

---

# ⭐ 9. **Tuple Packing & Unpacking**

### ✔ Packing:

```python
t = 1, 2, 3
```

### ✔ Unpacking:

```python
a, b, c = t
```

### ✔ Extended Unpacking:

```python
a, *b = (1, 2, 3, 4)
# a = 1
# b = [2, 3, 4]
```

---

# ⭐ 10. **Tuples with Mixed Data Types**

```python
t = (1, "Hello", 3.14, [10, 20])
```

Note: tuple is immutable but **list inside tuple is mutable**.

```python
t[3].append(30)
```

---

# ⭐ 11. **Nested Tuples**

```python
t = (1, (2, 3), (4, 5))
print(t[1][0])   # 2
```

---

# ⭐ 12. **Tuple vs List (Important for Interviews)**

| Feature    | Tuple      | List         |
| ---------- | ---------- | ------------ |
| Mutability | Immutable  | Mutable      |
| Speed      | Faster     | Slower       |
| Methods    | 2 methods  | Many methods |
| Memory     | Less       | More         |
| Use Case   | Fixed data | Dynamic data |

---

# ⭐ 13. **Why Use Tuples?**

✔ Faster execution
✔ Protects data from modification
✔ Can be used as dict keys (only if inner values also immutable)
✔ Useful in function returns
✔ Memory-efficient

---

# ⭐ 14. **Tuples as Dictionary Keys**

Allowed only if tuple elements are immutable:

```python
data = {(1, 2): "point"}
```

---

# ⭐ 15. **Tuple in Functions**

### ✔ Returning multiple values:

```python
def calc(a, b):
    return a+b, a-b, a*b

result = calc(10, 5)
```

### ✔ Using tuple unpacking:

```python
x, y, z = calc(10, 5)
```

---

# ⭐ 16. **Important Examples**

### ✔ Find min & max:

```python
t = (10, 50, 30)
print(min(t))
print(max(t))
```

### ✔ Count occurrences of each element:

```python
from collections import Counter
Counter((1, 2, 2, 3))
```

---

# ⭐ 17. **Short Summary (Quick Revision)**

* Tuple → ordered, immutable collection
* Created using parentheses or without
* Supports indexing, slicing, loops
* Only 2 methods: `count()`, `index()`
* Faster & memory efficient
* Useful when data should not change

# **4. Sets in Python**
---
A **set** is an **unordered**, **mutable**, and **unique-element** collection.

Example:

```python
s = {1, 2, 3}
```

---

# ⭐ 1. **Characteristics of Sets**

✔ **Unordered** → no indexing, no positions

✔ **Mutable** → can add/remove items

✔ **No duplicates allowed**

✔ **Heterogeneous allowed** (integers, strings, floats, tuples)

✔ **Can't store mutable items** (lists, dictionaries)

---

# ⭐ 2. **Creating Sets**

### ✔ Basic set:

```python
s = {1, 2, 3}
```

### ✔ Mixed data types:

```python
s = {1, "hello", 3.5}
```

### ✔ Empty set:

```python
s = set()   # correct
```

❌ Wrong:

```python
s = {}      # creates empty dictionary, not set
```

### ✔ Using set() constructor:

```python
s = set([1, 2, 3, 3])   # duplicates removed
```

---

# ⭐ 3. **Accessing Elements**

Sets **do not support indexing or slicing**.

❌ Not allowed:

```python
s[0]    # Error
```

✔ Iterate using loop:

```python
for x in s:
    print(x)
```

---

# ⭐ 4. **Basic Set Operations**

### ✔ Add element:

```python
s.add(5)
```

### ✔ Add multiple elements (update):

```python
s.update([6, 7, 8])
```

---

# ⭐ 5. **Remove Elements**

### ✔ remove() – gives error if element not found

```python
s.remove(2)
```

### ✔ discard() – no error if element missing

```python
s.discard(10)
```

### ✔ pop() – removes random element

```python
s.pop()
```

### ✔ clear() – empty the set

```python
s.clear()
```

---

# ⭐ 6. **Set Operations (Very Important)**

These follow mathematical set operations.

---

## 🔵 **1. Union (A ∪ B)**

Combine elements of both sets (no duplicates).

```python
A | B
A.union(B)
```

---

## 🔵 **2. Intersection (A ∩ B)**

Common elements.

```python
A & B
A.intersection(B)
```

---

## 🔵 **3. Difference (A - B)**

Elements in A but not in B.

```python
A - B
A.difference(B)
```

---

## 🔵 **4. Symmetric Difference (A Δ B)**

Elements in A or B but not in both.

```python
A ^ B
A.symmetric_difference(B)
```

---

# ⭐ 7. **Membership Test**

```python
5 in s     # True or False
```

---

# ⭐ 8. **Frozen Set (Immutable Set)**

A **frozenset** is like a set but **cannot be changed**.

```python
fs = frozenset([1, 2, 3])
```

Used when:
✔ You need a set as a dictionary key
✔ You want an immutable set

---

# ⭐ 9. **Set Methods List**

### Adding:

* `add()`
* `update()`

### Removing:

* `remove()`
* `discard()`
* `pop()`
* `clear()`

### Mathematical:

* `union()`
* `intersection()`
* `difference()`
* `symmetric_difference()`

### Checks:

* `issubset()`
* `issuperset()`
* `isdisjoint()`

---

# ⭐ 10. **Comparison Between Sets**

### ✔ Subset:

```python
A.issubset(B)
```

### ✔ Superset:

```python
A.issuperset(B)
```

### ✔ Disjoint:

```python
A.isdisjoint(B)   # True if no common elements
```

---

# ⭐ 11. **Important Examples**

### ✔ Remove duplicates from list:

```python
l = [1,2,2,3,3,4]
unique = set(l)
```

### ✔ Convert string to set of characters:

```python
set("hello")   # {'h','e','l','o'}
```

### ✔ Intersection of two lists:

```python
set(l1) & set(l2)
```

### ✔ Unique words from a sentence:

```python
sentence = "python is easy and python is powerful"
unique_words = set(sentence.split())
```

---

# ⭐ 12. **Applications of Sets**

✔ Removing duplicates
✔ Membership testing (very fast)
✔ Mathematical operations
✔ Data analytics, ML preprocessing
✔ Filtering unique values
✔ Intersection of datasets

---

# ⭐ 13. **Set vs List vs Tuple**

| Feature          | Set          | List          | Tuple      |
| ---------------- | ------------ | ------------- | ---------- |
| Order            | Unordered    | Ordered       | Ordered    |
| Mutable          | Yes          | Yes           | No         |
| Duplicates       | No           | Yes           | Yes        |
| Indexing         | No           | Yes           | Yes        |
| Speed for lookup | **Fastest**  | Slow          | Medium     |
| Use case         | Unique items | Dynamic items | Fixed data |

---

# ⭐ 14. **Short Summary (Quick Revision)**

* Set → unordered, mutable, no duplicates
* Created using `{}` or `set()`
* Supports mathematical operations (union, intersection...)
* No indexing or slicing
* Fast membership test
* `frozenset` is immutable version

# **5. Dictonary in Python**
---

A **dictionary** is a collection of data stored in **key–value pairs**.

Example:

```python
student = {"name": "Suhas", "age": 22}
```

---

# ⭐ 1. **Characteristics of Dictionary**

✔ **Ordered (Python 3.7+)**

✔ **Mutable** (can add/remove/update values)

✔ **Keys must be unique**

✔ **Keys must be immutable** (string, int, float, tuple)

✔ **Values can be anything** (list, tuple, set, dictionary, etc.)

---

# ⭐ 2. **Creating Dictionary**

### ✔ Using curly braces `{}`:

```python
d = {"name": "Ravi", "age": 25}
```

### ✔ Using dict() constructor:

```python
d = dict(name="Ravi", age=25)
```

### ✔ Empty dictionary:

```python
d = {}
```

### ✔ Dictionary with mixed data:

```python
d = {1: "one", "two": 2, 3.5: "float key"}
```

### ✔ Using tuple pairs:

```python
d = dict([("a", 1), ("b", 2)])
```

---

# ⭐ 3. **Accessing Values**

### ✔ Using key:

```python
d["name"]
```

### ✔ Using get() → avoids error if key not found:

```python
d.get("name")
d.get("salary", "Not Found")
```

---

# ⭐ 4. **Adding & Updating Items**

### ✔ Add new key-value:

```python
d["city"] = "Mumbai"
```

### ✔ Update existing key:

```python
d["age"] = 30
```

### ✔ update() method:

```python
d.update({"gender": "male"})
```

---

# ⭐ 5. **Removing Items**

### ✔ pop() → remove by key:

```python
d.pop("age")
```

### ✔ popitem() → removes **last inserted** key-value:

```python
d.popitem()
```

### ✔ del → remove key:

```python
del d["name"]
```

### ✔ clear() → empty dictionary:

```python
d.clear()
```

---

# ⭐ 6. **Dictionary Methods (Very Important)**

| Method      | Description                |
| ----------- | -------------------------- |
| `keys()`    | returns all keys           |
| `values()`  | returns all values         |
| `items()`   | returns (key, value) pairs |
| `get()`     | returns value safely       |
| `update()`  | merges / updates           |
| `pop()`     | remove by key              |
| `popitem()` | remove last item           |
| `clear()`   | empty dictionary           |

---

# ⭐ 7. **Looping Through Dictionary**

### ✔ Loop keys:

```python
for key in d:
    print(key)
```

### ✔ Loop values:

```python
for val in d.values():
    print(val)
```

### ✔ Loop both key & value:

```python
for key, val in d.items():
    print(key, val)
```

---

# ⭐ 8. **Dictionary Comprehension**

Create dictionary in one line.

### ✔ Example 1: Squares

```python
squares = {x: x*x for x in range(5)}
```

### ✔ Example 2: Filter dictionary

```python
d = {k: v for k, v in data.items() if v > 10}
```

---

# ⭐ 9. **Nested Dictionary**

Dictionary inside dictionary.

Example:

```python
student = {
    "name": "Amit",
    "marks": {"math": 90, "science": 88}
}
```

Access:

```python
student["marks"]["math"]
```

---

# ⭐ 10. **Modify Nested Dictionary**

```python
student["marks"]["math"] = 95
```

---

# ⭐ 11. **Important Concepts**

### ✔ Keys must be immutable:

Valid keys:

* int
* float
* string
* tuple

Invalid keys:

* list
* dictionary
* set

### ✔ Duplicate keys are not allowed:

```python
d = {"a": 1, "a": 2}
print(d)   # {'a': 2}
```

---

# ⭐ 12. **Useful Examples**

### ✔ 1. Convert two lists to dictionary:

```python
keys = ["name", "age"]
values = ["Ravi", 25]
d = dict(zip(keys, values))
```

### ✔ 2. Count frequency of characters:

```python
word = "banana"
freq = {}
for ch in word:
    freq[ch] = freq.get(ch, 0) + 1
```

### ✔ 3. Check if key exists:

```python
"name" in d
```

---

# ⭐ 13. **Dictionary vs List vs Set**

| Feature          | Dictionary | List   | Set           |
| ---------------- | ---------- | ------ | ------------- |
| Ordered          | Yes        | Yes    | No            |
| Mutable          | Yes        | Yes    | Yes           |
| Duplicate Keys   | No         | N/A    | N/A           |
| Duplicate Values | Yes        | Yes    | No            |
| Indexing         | No         | Yes    | No            |
| Stores           | Key-Value  | Values | Unique Values |

---

# ⭐ 14. **Quick Revision Summary**

* Dictionary = key–value pairs
* Keys must be unique & immutable
* Values can be anything
* Supports adding, updating, deleting
* Very important for real-world applications (JSON, APIs)
