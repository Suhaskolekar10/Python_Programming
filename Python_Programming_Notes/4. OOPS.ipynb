# **2. ENCAPSULATION IN PYTHON..**

## ⭐ **What is Encapsulation?**

Encapsulation means **binding data and methods into a single unit (class)** and **protecting the data from direct access**.

In simple terms:

> **Encapsulation hides the internal details of an object and only exposes the required parts.**

Python provides **three types of access modifiers** to implement encapsulation:

1. **Public**
2. **Protected** (`_variable`)
3. **Private** (`__variable`)

---

# ⭐ 1. **Public Members (Default)**

* Accessible from anywhere
* No restrictions

Example:

```python
class Student:
    def __init__(self, name):
        self.name = name  # public variable

obj = Student("Suhas")
print(obj.name)  # accessible
```

---

# ⭐ 2. **Protected Members (_variable)**

* Indicated by a **single underscore `_`**
* Should not be accessed outside class
* *Can be accessed*, but **not recommended** (by convention)

Example:

```python
class Student:
    def __init__(self):
        self._marks = 88   # protected

obj = Student()
print(obj._marks)  # allowed, but discouraged
```

---

# ⭐ 3. **Private Members (__variable)**

* Indicated by **double underscore `__`**
* Cannot be accessed directly outside the class
* Python performs **name mangling**:
  `__marks` → `_ClassName__marks`

Example:

```python
class Student:
    def __init__(self):
        self.__marks = 90  # private

obj = Student()
# print(obj.__marks)  ❌ Error
print(obj._Student__marks)  # ✔ Access using name-mangling
```

---

# ⭐ 4. **Why Encapsulation?**

✔ Data protection
✔ Control over data access
✔ Prevent accidental modification
✔ Implement getter & setter methods
✔ Improve security & maintainability

---

# ⭐ 5. **Getter and Setter Methods**

Used to access or modify private data safely.

### Example:

```python
class Person:
    def __init__(self, age):
        self.__age = age   # private

    # Getter
    def get_age(self):
        return self.__age

    # Setter
    def set_age(self, age):
        if age > 0:
            self.__age = age
        else:
            print("Invalid Age")
```

Usage:

```python
p = Person(21)
print(p.get_age())
p.set_age(25)
```

---

# ⭐ 6. **Encapsulation with Methods**

Methods can also be private.

```python
class Bank:
    def __init__(self):
        self.__balance = 1000

    def __update_balance(self):
        print("Balance updated!")

    def deposit(self, amount):
        self.__balance += amount
        self.__update_balance()
```

Here:

* `__balance` → private attribute
* `__update_balance()` → private method

---

# ⭐ 7. **Strong Encapsulation Using Private Members**

Example:

```python
class ATM:
    def __init__(self):
        self.__pin = "1234"

    def get_pin(self):
        return "Access Denied"  # cannot reveal pin

    def set_pin(self, pin):
        if len(pin) == 4:
            self.__pin = pin
        else:
            print("Invalid PIN")
```

This protects sensitive data (like PIN).

---

# ⭐ 8. **Encapsulation vs Abstraction (Important)**

| Concept           | Meaning                            | Focus                     |
| ----------------- | ---------------------------------- | ------------------------- |
| **Encapsulation** | Hiding data using access modifiers | Protect data              |
| **Abstraction**   | Hiding complexity                  | Show only essential parts |

---

# ⭐ 9. **Example Showing All Access Levels**

```python
class Demo:
    a = 10        # public
    _b = 20       # protected
    __c = 30      # private

obj = Demo()
print(obj.a)       # public
print(obj._b)      # protected (accessible but not recommended)
# print(obj.__c)   # private: error
print(obj._Demo__c)  # name-mangled
```

---

# ⭐ 10. **Real-Life Example: Bank Account**

```python
class BankAccount:
    def __init__(self, balance):
        self.__balance = balance  # private

    def deposit(self, amount):
        self.__balance += amount

    def get_balance(self):
        return self.__balance
```

User can access balance only via `get_balance()` → Safe!

---

# ⭐ QUICK REVISION

* Encapsulation = **protecting data**
* Public → accessible anywhere
* Protected → `_var` (convention)
* Private → `__var` (name mangling)
* Use **getters/setters** to access private data
* Improves **security, maintainability, safety**

# **3. ABSTRACTION IN PYTHON**
---
## ⭐ **What is Abstraction?**

**Abstraction** means **hiding internal implementation details** and showing **only essential information** to the user.

It focuses on **“what it does”** rather than **“how it does”**.

### Real-Life Example:

* You drive a car using:

  * Steering
  * Brake
  * Accelerator

You **don’t care** how the engine works internally.
That’s **abstraction**.

---

# ⭐ **Why Abstraction?**

✔ Reduces complexity
✔ Improves readability
✔ Prevents misuse
✔ Helps in designing cleaner, modular code
✔ Allows programmers to hide sensitive logic

---

# ⭐ **How is Abstraction done in Python?**

Python uses **Abstract Classes** and **Interfaces** (via `abc` module):

```python
from abc import ABC, abstractmethod
```

An abstract class:

* Cannot be instantiated
* Can contain abstract + normal methods
* Forces child classes to give implementation

---

# ⭐ 1. **Abstract Class**

Example:

```python
from abc import ABC, abstractmethod

class Animal(ABC):

    @abstractmethod
    def sound(self):
        pass

class Dog(Animal):
    def sound(self):
        print("Bark!")

class Cat(Animal):
    def sound(self):
        print("Meow!")

d = Dog()
d.sound()  # Bark!
```

### Key Points:

* `Animal` → abstract class
* `sound()` → abstract method (must be overridden)
* Cannot create `Animal()` object

---

# ⭐ 2. **Abstract Class with Normal Methods**

```python
from abc import ABC, abstractmethod

class Vehicle(ABC):

    def describe(self):
        print("This is a vehicle")

    @abstractmethod
    def start(self):
        pass

class Car(Vehicle):
    def start(self):
        print("Car starts with a key")
```

Here:

* `describe()` is normal method
* `start()` must be implemented by child class

---

# ⭐ 3. **Abstraction using Interfaces**

Python doesn’t have interfaces like Java,
but we can create interface-like classes using abstract methods only.

```python
from abc import ABC, abstractmethod

class PaymentGateway(ABC):

    @abstractmethod
    def pay(self):
        pass
```

---

# ⭐ 4. **Why Abstract Classes Are Useful?**

Abstract classes help enforce **standards**.

Example: Suppose you're making a Payment System:

```python
class Payment(ABC):

    @abstractmethod
    def process_payment(self):
        pass
```

Now, every payment type (UPI, Card, NetBanking) must implement it.

This ensures all payment classes follow the **same rule**.

---

# ⭐ 5. **Abstraction vs Encapsulation**

| Feature    | Encapsulation            | Abstraction                        |
| ---------- | ------------------------ | ---------------------------------- |
| Purpose    | Protect data             | Hide complexity                    |
| Hidden     | Data                     | Internal implementation            |
| Focus      | **How data is accessed** | **What methods do**                |
| Tools Used | Private, Protected vars  | Abstract classes, abstract methods |

---

# ⭐ 6. **Example with Real-Life Logic Hidden**

```python
from abc import ABC, abstractmethod

class ATM(ABC):

    @abstractmethod
    def authenticate_user(self):
        pass

    @abstractmethod
    def withdraw(self, amount):
        pass

class SBI(ATM):
    def authenticate_user(self):
        print("Authenticating user with SBI DB...")

    def withdraw(self, amount):
        print(f"Withdrawing ₹{amount} from SBI ATM")
```

User only calls:

```python
atm = SBI()
atm.authenticate_user()
atm.withdraw(1000)
```

But does **not** know internally how authentication works → abstraction.

---

# ⭐ 7. **Abstract Class with Constructor**

```python
from abc import ABC, abstractmethod

class Shape(ABC):
    def __init__(self, color):
        self.color = color

    @abstractmethod
    def area(self):
        pass

class Circle(Shape):
    def area(self):
        return 3.14 * 5 * 5
```

Even constructors can be part of abstract classes.

---

# ⭐ 8. **Multiple Abstract Methods**

```python
class Laptop(ABC):

    @abstractmethod
    def keyboard(self):
        pass

    @abstractmethod
    def screen(self):
        pass

    @abstractmethod
    def processor(self):
        pass
```

Child class must implement **all** abstract methods.

---

# ⭐ Quick Summary

* Abstraction hides internal logic
* Achieved using **abstract classes & abstract methods**
* Cannot create objects of abstract class
* Child classes must implement abstract methods
* Helps write scalable, clean code

# **4. POLYMORPHISM IN PYTHON..**

## ⭐ What is Polymorphism?

**Polymorphism = “Many Forms”**

In OOP, polymorphism means:

> **The same function/method/operator behaves differently depending on the context.**

Example (simple idea):

* `+` works as addition: `3 + 4`
* `+` works as concatenation: `"Suhas" + "Python"`
* `+` works as list merge: `[1,2] + [3,4]`

Same symbol → Different behaviour.

---

# ⭐ Types of Polymorphism in Python

1. **Method Overloading** (not directly supported, but achievable)
2. **Method Overriding** (supported)
3. **Operator Overloading**
4. **Duck Typing**

We will learn each one with simple examples.

---

# 🔵 1. **Method Overriding** (REAL polymorphism)

Method overriding happens when a **child class gives its own version of a method** already present in parent class.

```python
class Animal:
    def sound(self):
        print("Animal makes sound")

class Dog(Animal):
    def sound(self):
        print("Dog barks")

class Cat(Animal):
    def sound(self):
        print("Cat meows")

# Polymorphism
for animal in [Dog(), Cat()]:
    animal.sound()
```

Output:

```
Dog barks
Cat meows
```

👉 Same method name, different behaviour depending on object type.

---

# 🔵 2. **Method Overloading** (Python-style)

Python **does NOT support overloading like Java**, but we can mimic it using:

### ✔ Default Arguments

### ✔ Variable Arguments (`*args`)

Example:

```python
class Add:
    def sum(self, a=0, b=0, c=0):
        return a + b + c

obj = Add()

print(obj.sum(5))        # 5
print(obj.sum(5, 10))    # 15
print(obj.sum(5, 10, 20))# 35
```

### Python chooses based on **number of arguments** you pass.

---

# 🔵 3. **Operator Overloading**

Python allows you to redefine how operators work for user-defined classes using **special methods** (magic methods).

Examples:

| Operator | Magic Method |
| -------- | ------------ |
| `+`      | `__add__()`  |
| `-`      | `__sub__()`  |
| `*`      | `__mul__()`  |
| `==`     | `__eq__()`   |

Example:

```python
class Student:
    def __init__(self, marks):
        self.marks = marks

    def __add__(self, other):
        return self.marks + other.marks

s1 = Student(50)
s2 = Student(70)

print(s1 + s2)  # 120
```

👉 `+` operator is redefined to add marks of students.

---

# 🔵 4. **Duck Typing** (Python’s most powerful form)

Python’s philosophy:

> **“If it looks like a duck and quacks like a duck — it’s a duck.”**

Meaning:

* If two objects have the same method name
* Python allows them to be used interchangeably
* Based on **behavior**, not class type

Example:

```python
class PyCharm:
    def execute(self):
        print("Running code...")

class VSCode:
    def execute(self):
        print("Compiling and running...")

def runIDE(ide):
    ide.execute()

runIDE(PyCharm())   # Works
runIDE(VSCode())    # Works
```

👉 Python doesn’t care about the class, it cares whether `execute()` exists.

---

# ⭐ Polymorphism with Functions

A single function can work for different types.

```python
print(len("Suhas"))        # length of string
print(len([1,2,3,4]))      # length of list
print(len({"a":1, "b":2})) # length of dictionary
```

`len()` behaves differently depending on the object passed.

---

# ⭐ Polymorphism with Class Method

```python
class Bird:
    def intro(self):
        print("There are many types of birds")

    def flight(self):
        print("Most birds can fly")

class Penguin(Bird):
    def flight(self):
        print("Penguins cannot fly")
```

Calling:

```python
obj1 = Bird()
obj2 = Penguin()

obj1.flight()
obj2.flight()
```

---

# ⭐ Polymorphism in Loops

```python
class India:
    def capital(self):
        print("New Delhi")

class USA:
    def capital(self):
        print("Washington DC")

for country in (India(), USA()):
    country.capital()
```

Each object responds in its own way → polymorphism.

---

# ⭐ Quick Formula to Remember

```
Polymorphism = Same Name + Different Behaviour
```

---

# ⭐ Summary Chart

| Operator Overloading | ✔ Yes                | Redefining operators using magic methods    |
| Duck Typing          | ✔ Yes                | Works if methods exist, regardless of class |

# **5. INHERITANCE IN PYTHON**
---

## ⭐ **What is Inheritance?**

Inheritance allows one class (child/subclass) to **acquire the properties and methods** of another class (parent/superclass).

It helps achieve:

* **Code reusability**
* **Less duplication**
* **Better organization**
* **Logical relationship between classes**

---

# ⭐ 1. **Basic Inheritance**

```python
class Parent:
    def display(self):
        print("Parent class method")

class Child(Parent):
    pass

obj = Child()
obj.display()
```

✔ The `Child` inherits `display()` from `Parent`.

---

# ⭐ 2. **Types of Inheritance in Python**

Python supports **5 types**:

1. **Single inheritance**
2. **Multilevel inheritance**
3. **Multiple inheritance**
4. **Hierarchical inheritance**
5. **Hybrid inheritance**

Let’s understand each.

---

# 🟦 **1. Single Inheritance**

One parent → One child.

```python
class A:
    def showA(self):
        print("Class A")

class B(A):
    def showB(self):
        print("Class B")
```

---

# 🟦 **2. Multilevel Inheritance**

Grandparent → Parent → Child.

```
A → B → C
```

```python
class A:
    def showA(self):
        print("A")

class B(A):
    def showB(self):
        print("B")

class C(B):
    def showC(self):
        print("C")
```

✔ Class C inherits from both A and B.

---

# 🟦 **3. Multiple Inheritance**

Child inherits from **more than one parent**.

```python
class A:
    def methodA(self):
        print("A")

class B:
    def methodB(self):
        print("B")

class C(A, B):
    pass

obj = C()
obj.methodA()
obj.methodB()
```

Python solves conflicts using **MRO (Method Resolution Order)**.

---

# 🟦 **4. Hierarchical Inheritance**

One parent → Multiple children.

```python
class Parent:
    def show(self):
        print("Parent")

class Child1(Parent):
    pass

class Child2(Parent):
    pass
```

---

# 🟦 **5. Hybrid Inheritance**

Combination of multiple types.

Example:
Hierarchical + Multiple = Hybrid.

---

# ⭐ **Constructor in Inheritance**

Parent’s constructor runs **automatically** unless overridden.

```python
class Parent:
    def __init__(self):
        print("Parent constructor")

class Child(Parent):
    pass

obj = Child()
```

---

# ⭐ **Calling Parent Constructor (super())**

When child has its own constructor:

```python
class Parent:
    def __init__(self):
        print("Parent constructor")

class Child(Parent):
    def __init__(self):
        super().__init__()
        print("Child constructor")
```

✔ `super()` calls the parent class constructor.

---

# ⭐ **Method Overriding (Important in Inheritance)**

Child class can override parent class method.

```python
class Parent:
    def greet(self):
        print("Hello from Parent")

class Child(Parent):
    def greet(self):
        print("Hello from Child")
```

✔ Most used with polymorphism.

---

# ⭐ **super() with Methods**

```python
class Parent:
    def message(self):
        print("Parent Message")

class Child(Parent):
    def message(self):
        super().message()
        print("Child Message")
```

---

# ⭐ **MRO (Method Resolution Order)**

Used when multiple parents exist.

Python uses:

> **C3 Linearization / Depth-first left-to-right order**

Check MRO:

```python
print(ClassName.mro())
```

---

# ⭐ **Private Members in Inheritance**

Private members (`__var`) are **not inherited** directly.

```python
class A:
    __secret = 10  # private

class B(A):
    pass

obj = B()
# print(obj.__secret)  ❌ Not accessible
```

But accessible via name-mangling:

```python
print(obj._A__secret)
```

(Not recommended.)

---

# ⭐ **Real-Life Example of Inheritance**

```
Vehicle (Parent)
    ↳ Car (Child)
    ↳ Bike (Child)
```

```python
class Vehicle:
    def start(self):
        print("Starting vehicle")

class Car(Vehicle):
    def start(self):
        print("Car starts with a key")

class Bike(Vehicle):
    def start(self):
        print("Bike starts with button")
```

---

# ⭐ **When to Use Inheritance?**

Use inheritance when:

✔ "Is-A" relationship
✔ Code reuse
✔ Standard behavior shared among classes

Example:

* Car **is a** Vehicle
* Dog **is an** Animal
* SavingsAccount **is a** BankAccount
