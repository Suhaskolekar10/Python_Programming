# **1. FILE HANDLING IN PYTHON**

File handling allows Python to **create, read, write, update, and delete** files.

Python uses the built-in function **`open()`** to work with files.

---

# ⭐ 1. **Opening a File**

### **Syntax:**

```python
open(filename, mode)
```

### ✔ Common Modes

| Mode   | Meaning       | Description                              |
| ------ | ------------- | ---------------------------------------- |
| `"r"`  | Read          | Default mode, file must exist            |
| `"w"`  | Write         | Creates new file OR overwrites existing  |
| `"a"`  | Append        | Writes at end of file, preserves content |
| `"x"`  | Create        | Creates new file, error if exists        |
| `"b"`  | Binary        | Used for images, PDFs, etc.              |
| `"t"`  | Text          | Default text mode                        |
| `"r+"` | Read + Write  | File must exist                          |
| `"w+"` | Write + Read  | Overwrites file                          |
| `"a+"` | Append + Read | Writes at end                            |

---

# ⭐ 2. **Closing a File**

Always needed to avoid memory leaks.

```python
f = open("data.txt", "r")
f.close()
```

But best practice:

---

# ⭐ 3. **Using `with` (Automatic Closing)** ✔ BEST METHOD

```python
with open("data.txt", "r") as f:
    content = f.read()
```

File closes automatically after block ends.

---

# ⭐ 4. **Reading Files**

### ⭐ **1. read()** → Reads entire file

```python
with open("data.txt", "r") as f:
    data = f.read()
```

---

### ⭐ **2. readline()** → Reads one line

```python
line = f.readline()
```

---

### ⭐ **3. readlines()** → Reads all lines into list

```python
lines = f.readlines()
```

---

### ⭐ Read file line by line (best)

```python
for line in f:
    print(line)
```

---

# ⭐ 5. **Writing to Files**

### ✔ Overwrite file (write mode):

```python
with open("data.txt", "w") as f:
    f.write("Hello World")
```

### ✔ Write multiple lines:

```python
f.writelines(["A\n", "B\n", "C\n"])
```

---

# ⭐ 6. **Appending to Files**

```python
with open("data.txt", "a") as f:
    f.write("\nNew line added!")
```

---

# ⭐ 7. **Working With Binary Files**

Used for images, audio, video, PDF, etc.

### ✔ Reading binary:

```python
with open("photo.jpg", "rb") as f:
    data = f.read()
```

### ✔ Writing binary:

```python
with open("copy.jpg", "wb") as f:
    f.write(data)
```

---

# ⭐ 8. **File Pointer Position**

### ✔ Get current position:

```python
f.tell()
```

### ✔ Move pointer:

```python
f.seek(offset)
```

Example:

```python
f.seek(0)  # go to beginning
```

---

# ⭐ 9. **Check if File Exists**

Using `os` module:

```python
import os
os.path.exists("data.txt")
```

---

# ⭐ 10. **Delete a File**

```python
import os
os.remove("data.txt")
```

---

# ⭐ 11. **Rename File**

```python
os.rename("old.txt", "new.txt")
```

---

# ⭐ 12. **Create Folder / Directory**

```python
os.mkdir("myfolder")
```

---

# ⭐ 13. **List Files in Folder**

```python
os.listdir()
```

---

# ⭐ 14. **Exception Handling in File Handling**

To avoid crashes when file is missing:

```python
try:
    f = open("data.txt", "r")
    print(f.read())
except FileNotFoundError:
    print("File not found!")
```

---

# ⭐ 15. **Important Concepts**

### ✔ Text mode vs Binary mode

* `"t"` → text (default)
* `"b"` → binary

### ✔ Overwriting:

`"w"` always erases previous content.

### ✔ `with` block is safest for file operations.

---

# ⭐ 16. **Examples**

### ✔ Example 1: Copy content from one file to another

```python
with open("source.txt", "r") as s, open("dest.txt", "w") as d:
    d.write(s.read())
```

---

### ✔ Example 2: Count lines, words, characters

```python
with open("file.txt", "r") as f:
    data = f.read()

lines = data.count("\n") + 1
words = len(data.split())
chars = len(data)
```

---

### ✔ Example 3: Reverse content of a file

```python
with open("data.txt", "r") as f:
    d = f.read()

with open("rev.txt", "w") as f:
    f.write(d[::-1])
```

---

# ⭐ 17. Quick Revision Notes

* `open()` → open file
* `"r"` → read, `"w"` → write, `"a"` → append
* Use `with` for auto-close
* `read()`, `readline()`, `readlines()`
* `write()`, `writelines()`
* Binary mode uses `"rb"` / `"wb"`
* `seek()` moves pointer
* `tell()` shows pointer position
* Use `os.remove()` to delete files

# **2. ERROR HANDLING IN PYTHON..**

When a program encounters an **error**, Python stops execution and shows a message called an **exception**.

To prevent crashing, we use **exception handling**.

---

# ⭐ 1. **What is an Exception?**

An **exception** is a runtime error that stops the program.

Examples:

* `ZeroDivisionError`
* `ValueError`
* `TypeError`
* `FileNotFoundError`

---

# ⭐ 2. **Why Use Error Handling?**

✔ Prevent program from crashing
✔ Show meaningful error messages
✔ Let the program continue
✔ Handle unexpected user input

---

# ⭐ 3. **Basic Syntax (try–except)**

```python
try:
    risky_code
except:
    handle_error
```

### Example:

```python
try:
    x = 10 / 0
except:
    print("Error occurred!")
```

---

# ⭐ 4. **Handling Specific Exceptions**

Always better than a general `except`.

```python
try:
    a = int("abc")
except ValueError:
    print("Invalid integer!")
```

---

# ⭐ 5. **Multiple Except Blocks**

```python
try:
    n = int(input("Enter number: "))
    print(10 / n)
except ValueError:
    print("Enter numbers only!")
except ZeroDivisionError:
    print("Cannot divide by zero!")
```

---

# ⭐ 6. **Using else Block**

Runs only if **no exception occurs**.

```python
try:
    f = open("data.txt", "r")
except FileNotFoundError:
    print("File missing.")
else:
    print("File opened successfully.")
    f.close()
```

---

# ⭐ 7. **Using finally Block**

Runs **always**, with or without exception.

Used to close resources like files, DB, etc.

```python
try:
    f = open("data.txt", "r")
except:
    print("Error!")
finally:
    print("Closing program...")
```

---

# ⭐ 8. **Try–Except–Else–Finally (Full Structure)**

```python
try:
    # risky code
except:
    # runs if error occurs
else:
    # runs if no error
finally:
    # runs always
```

---

# ⭐ 9. **Raise Your Own Exception**

Use `raise` to manually trigger an error.

```python
age = -1

if age < 0:
    raise ValueError("Age cannot be negative")
```

---

# ⭐ 10. **Custom Exceptions (User-Defined)**

Used when default exceptions don't fit.

```python
class MyError(Exception):
    pass

try:
    raise MyError("Custom error occurred!")
except MyError as e:
    print(e)
```

---

# ⭐ 11. **Common Python Exceptions**

| Exception           | Meaning                   |
| ------------------- | ------------------------- |
| `ZeroDivisionError` | Dividing by 0             |
| `ValueError`        | Wrong value type          |
| `TypeError`         | Wrong data type           |
| `IndexError`        | Index out of range        |
| `KeyError`          | Missing key in dictionary |
| `FileNotFoundError` | File not found            |
| `NameError`         | Variable not defined      |
| `AttributeError`    | Invalid property access   |
| `MemoryError`       | Out of memory             |

---

# ⭐ 12. **Example: Handling Multiple Errors Together**

```python
try:
    a = int(input("Enter number: "))
    b = 10 / a
except (ValueError, ZeroDivisionError):
    print("Check your input!")
```

---

# ⭐ 13. **Nested Try-Except**

```python
try:
    a = int(input("Enter number: "))
    try:
        result = 10 / a
    except ZeroDivisionError:
        print("Cannot divide by zero")
except ValueError:
    print("Enter valid integers only")
```

---

# ⭐ 14. **Practical Example: Safe Input**

```python
while True:
    try:
        n = int(input("Enter a number: "))
        break
    except ValueError:
        print("Invalid input, try again!")
```

---

# ⭐ QUICK REVISION

* `try`: code that may cause error
* `except`: handles the error
* `else`: runs when no error
* `finally`: runs always
* `raise`: manually throw error
* Good practice → catch specific exceptions

# **3. MODULES AND PACKAGES IN PYTHON..**

Python supports code reusability and modular programming using:

✔ **Modules**
✔ **Packages**

---

# 🟦 1. **What is a Module?**

A **module** is simply a Python file (`.py`) that contains:

* Variables
* Functions
* Classes

Example:
`math.py`, `calculator.py`, `user_utils.py`

---

# 🟦 2. **Why Use Modules?**

✔ Reusable code
✔ Break big programs into smaller pieces
✔ Easy maintenance
✔ Avoid code duplication

---

# 🟦 3. **Creating a Module**

Create a file:
`mymodule.py`

```python
def greet(name):
    return f"Hello, {name}"

x = 10
```

Use the module:

```python
import mymodule
print(mymodule.greet("Suhas"))
print(mymodule.x)
```

---

# 🟦 4. **Types of Import**

### ✔ 1. import module

```python
import math
print(math.sqrt(16))
```

---

### ✔ 2. from module import name

```python
from math import sqrt
print(sqrt(16))
```

---

### ✔ 3. Import multiple items

```python
from math import sin, cos
```

---

### ✔ 4. Import all items (NOT recommended)

```python
from math import *
```

---

### ✔ 5. Import with alias

```python
import math as m
print(m.pi)
```

---

# 🟦 5. **Built-In Modules in Python**

Some examples:

* `math`
* `random`
* `datetime`
* `os`
* `sys`
* `json`
* `csv`
* `collections`

---

# 🟩 6. **What is a Package?**

A **package** is a collection of **modules** stored inside a folder.

A folder becomes a package if it contains:

```
__init__.py
```

Example package structure:

```
mypackage/
    __init__.py
    math_ops.py
    string_ops.py
```

---

# 🟩 7. **Creating a Package**

### Step 1: Folder structure

```
utilities/
    __init__.py
    calculator.py
    formatter.py
```

`__init__.py` makes it a package.

---

### Step 2: Write code in modules

**calculator.py**

```python
def add(a, b):
    return a + b
```

---

### Step 3: Import package module

```python
from utilities.calculator import add
print(add(5, 3))
```

---

# 🟩 8. ****init**.py (Very Important)**

* Marks directory as a package
* Can initialize variables
* Can import functions

Example:

```python
from .calculator import add
```

---

Example:

```
project/
    data/
        __init__.py
        file_utils.py
    math/
        __init__.py
        algebra.py
```

Importing subpackages:

```python
from project.math.algebra import solve
```

# 🟦 12. **Advantages of Modules & Packages**

✔ Code reuse
✔ Cleaner structure
✔ Easy debugging
✔ Collaborating in large projects
✔ Prevent name conflicts using namespaces

# 🟦 14. **Difference Between Module & Package**

| Feature    | Module                        | Package                 |
| ---------- | ----------------------------- | ----------------------- |
| Meaning    | Single .py file               | Collection of modules   |
| Structure  | One file                      | Folder with sub-folders |
| Contains   | Variables, functions, classes | Modules, sub-packages   |
| Identifier | No **init**.py needed         | Needs **init**.py       |
