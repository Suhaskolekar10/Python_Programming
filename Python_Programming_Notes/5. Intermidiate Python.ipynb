# **1. ITERATORS & GENERATORS**

---

# 🟦 PART 1: **Iterators in Python**

## ⭐ What is an Iterator?

An **iterator** is an object that allows you to **iterate (loop)** through values **one at a time**.

An object is an iterator if it has:

* `__iter__()` → returns the iterator object
* `__next__()` → returns the next value

When values are finished, `StopIteration` is raised.

---

## ⭐ Example of Built-in Iterators

Python objects like list, tuple, string, dict are **iterables**, not iterators.

We convert iterables → iterator using `iter()`.

```python
numbers = [1, 2, 3]
it = iter(numbers)

print(next(it))  # 1
print(next(it))  # 2
print(next(it))  # 3
```

---

## ⭐ Custom Iterator Example

```python
class Count:
    def __init__(self, n):
        self.n = n
        self.i = 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.i <= self.n:
            val = self.i
            self.i += 1
            return val
        else:
            raise StopIteration

obj = Count(5)

for num in obj:
    print(num)
```

Output:

```
1
2
3
4
5
```

---

# ⭐ Why Iterators?

✔ They save memory
✔ Allow looping over large datasets
✔ You get values one by one, not all at once

---

# 🟩 PART 2: **Generators in Python**

A **generator** is a special type of iterator
created using the `yield` keyword.

### Difference:

* **Iterator** → Class + `__iter__()` + `__next__()`
* **Generator** → function + `yield`
* Generator is **simpler and lightweight**

---

## ⭐ Example: Generator Function

```python
def mygen():
    yield 1
    yield 2
    yield 3

g = mygen()

print(next(g))  # 1
print(next(g))  # 2
print(next(g))  # 3
```

---

# ⭐ How generator works internally?

Every time `yield` is hit:

* Function **pauses**
* State is saved
* Next call resumes from the same point

---

# ⭐ Generator with Loop

```python
def counter(n):
    for i in range(1, n+1):
        yield i

for num in counter(5):
    print(num)
```

---

# ⭐ Infinite Sequence Generator

```python
def infinite_counter():
    i = 1
    while True:
        yield i
        i += 1
```

✔ No memory overload
✔ Infinite values on demand

---

# ⭐ Generator Expressions

Similar to list comprehensions but use `()` instead of `[]`

```python
gen = (x*x for x in range(5))

for i in gen:
    print(i)
```

Uses **much less memory** than list comprehension.

---

# 🟦 Iterators vs Generators — Difference Table

| Feature          | Iterator | Generator             |
| ---------------- | -------- | --------------------- |
| Creation         | Class    | Function with `yield` |
| Memory Use       | High     | Very low              |
| Implementation   | Complex  | Simple                |
| Infinite Loop    | Hard     | Easy                  |
| Built-in support | No       | Yes (`yield`)         |

---

# 🟦 yield vs return

| return               | yield                              |
| -------------------- | ---------------------------------- |
| Ends function        | Pauses function                    |
| Returns single value | Returns multiple values one-by-one |
| No saved state       | Saves state                        |
| Cannot resume        | Can resume                         |

---

# ⭐ Real-Life Example: Reading a Large File

**Generator: Best way to read big files.**

```python
def read_file():
    with open("bigdata.txt") as f:
        for line in f:
            yield line
```

✔ Reads file **line by line**
✔ Saves memory

---

# ⭐ Another Example: Fibonacci Generator

```python
def fibonacci(n):
    a, b = 0, 1
    for _ in range(n):
        yield a
        a, b = b, a + b

for x in fibonacci(10):
    print(x)
```

---

# ⭐ Summary

### ITERATOR:

* Uses `__iter__()` and `__next__()`
* Holds its state
* Needs custom class

### GENERATOR:

* Uses `yield`
* Automatically creates iterator
* Very memory efficient
* Best for large data, streaming, infinite sequences

# **2. DECORATORS IN PYTHON**
---
## ⭐ What is a Decorator?

**Decorators allow you to add new functionality to an existing function *without changing its code*.**

They take a function as input → add extra features → return a new function.

### Think of a decorator as:

> **A wrapper around another function.**

---

# ⭐ Decorators Basics

In Python:

* Functions can be passed as arguments
* Functions can return other functions

This allows decorators to work.

---

# 🧩 **Simple Decorator Example**

```python
def my_decorator(func):
    def wrapper():
        print("Before function execution")
        func()
        print("After function execution")
    return wrapper

@my_decorator
def say_hello():
    print("Hello!")

say_hello()
```

### Output:

```
Before function execution
Hello!
After function execution
```

### Here:

* `@my_decorator` modifies `say_hello()`
* `wrapper()` adds extra code before & after the original function

---

# ⭐ How Decorators Work Internally

This:

```python
@my_decorator
def say_hello():
    print("Hello")
```

is the same as:

```python
say_hello = my_decorator(say_hello)
```

---

# ⭐ Decorator with Arguments

When the function has parameters:

```python
def smart_divide(func):
    def wrapper(a, b):
        print("Dividing", a, "by", b)
        if b == 0:
            print("Cannot divide by zero!")
            return
        return func(a, b)
    return wrapper

@smart_divide
def divide(a, b):
    print(a / b)

divide(10, 2)
divide(5, 0)
```

---

# ⭐ Decorators That Accept Arguments → (3-level wrapping)

Example: A decorator that takes input itself:

```python
def repeat(n):
    def decorator(func):
        def wrapper():
            for _ in range(n):
                func()
        return wrapper
    return decorator

@repeat(3)
def greet():
    print("Hello")

greet()
```

Output:

```
Hello
Hello
Hello
```

> The function runs 3 times because the decorator took an argument.

---

# ⭐ Decorators with `*args` and `**kwargs`

Useful when the decorated function has unknown number of arguments.

```python
def logger(func):
    def wrapper(*args, **kwargs):
        print("Arguments:", args, kwargs)
        return func(*args, **kwargs)
    return wrapper

@logger
def add(a, b):
    return a + b

print(add(5, 3))
```

---

# ⭐ Using `functools.wraps` (Best Practice)

Decorators override metadata (like function name).
Use `wraps` to preserve it.

```python
from functools import wraps

def my_decorator(func):
    @wraps(func)
    def wrapper():
        print("Before")
        func()
        print("After")
    return wrapper
```

---

# 🎯 Real-Life Use Cases of Decorators

### ✔ 1. Logging

```python
def log(func):
    def wrapper(*args):
        print(f"Running {func.__name__}")
        return func(*args)
    return wrapper
```

---

### ✔ 2. Authorization

```python
def authorize(func):
    def wrapper(user):
        if user != "admin":
            print("Access denied")
            return
        return func(user)
    return wrapper

@authorize
def panel(user):
    print("Welcome admin")
```

---

### ✔ 3. Time Measurement

```python
import time

def timer(func):
    def wrapper(*args):
        start = time.time()
        result = func(*args)
        print("Time:", time.time() - start)
        return result
    return wrapper
```

---

### ✔ 4. Caching (Memoization)

```python
def cache(func):
    memo = {}
    def wrapper(n):
        if n in memo:
            return memo[n]
        memo[n] = func(n)
        return memo[n]
    return wrapper

@cache
def factorial(n):
    return 1 if n==0 else n * factorial(n-1)
```

---

# ⭐ Decorating Class Methods

```python
def debug(func):
    def wrapper(*args):
        print("Calling:", func.__name__)
        return func(*args)
    return wrapper

class Person:
    @debug
    def greet(self):
        print("Hello")
```

---

# ⭐ Multiple Decorators on One Function

```python
@decorator1
@decorator2
def function():
    pass
```

Execution order:

```
function → decorator2 → decorator1
```

| `wrapper()`         | Inner function that adds extra code     |
| `*args`, `**kwargs` | For variable arguments                  |

# **3. Context Managers in Python (`with` Statement)**

---

## **1. What is a Context Manager?**

A **context manager** is a Python construct that **handles setup and cleanup automatically** when working with resources like:

* Files
* Network connections
* Database connections
* Locks/threads
* Temporary directories
* Opening/closing anything safely

The **with statement** is used to **enter** and **exit** such contexts without manually writing cleanup code.

---

# ⭐ 2. Why use Context Managers?

### Without a context manager:

```python
file = open("data.txt", "r")
try:
    content = file.read()
finally:
    file.close()
```

### With a context manager:

```python
with open("data.txt", "r") as file:
    content = file.read()
```

✔ Automatically opens the file
✔ Automatically closes it after the block ends
✔ Even if error occurs, file is closed safely

➡ **Cleaner, safer, more Pythonic**

---

# ⭐ 3. How `with` Statement Works Internally

A context manager must implement two magic methods:

| Method                                           | Purpose                                                       |
| ------------------------------------------------ | ------------------------------------------------------------- |
| `__enter__(self)`                                | Code executed when entering the context (`with` block starts) |
| `__exit__(self, exc_type, exc_value, traceback)` | Code executed when leaving the context (cleanup)              |

---

# ⭐ 4. Creating Your Own Context Manager (Class-based)

### Example 1: Simple Context Manager

```python
class MyContext:
    def __enter__(self):
        print("Entering the block")
        return "Resource"

    def __exit__(self, exc_type, exc_value, traceback):
        print("Exiting the block")

with MyContext() as r:
    print("Inside block:", r)
```

**Output:**

```
Entering the block
Inside block: Resource
Exiting the block
```

---

# ⭐ 5. Handling Errors Inside `with`

If an exception occurs inside the block, `__exit__` receives details.

Example:

```python
class Demo:
    def __enter__(self):
        print("Start")
        return self

    def __exit__(self, exc_type, exc_value, traceback):
        print("Exception handled:", exc_type)
        return True  # prevents crashing

with Demo():
    x = 1 / 0   # error
```

✔ Program doesn’t crash
✔ Error is handled inside context manager

---

# ⭐ 6. Context Manager Using **Decorator** (`contextlib`)

Python provides an easier way:

```python
from contextlib import contextmanager

@contextmanager
def open_file(path, mode):
    f = open(path, mode)
    try:
        yield f   # this becomes `as f`
    finally:
        f.close()

with open_file("test.txt", "w") as file:
    file.write("Hello!")
```

This is simpler than writing a full class.

---

# ⭐ 7. Real-World Examples of Context Managers

### ✔ Opening files

```python
with open("data.txt") as f:
    print(f.read())
```

### ✔ Working with database connections

```python
with db.connect() as conn:
    conn.execute("SELECT * FROM users")
```

### ✔ Lock management in multithreading

```python
with lock:
    shared_data += 1
```

### ✔ Temporary changes (like changing directory)

```python
import os
from contextlib import contextmanager

@contextmanager
def change_dir(path):
    old = os.getcwd()
    os.chdir(path)
    try:
        yield
    finally:
        os.chdir(old)

with change_dir("/tmp"):
    print(os.listdir())
```

---

# ⭐ 8. When to Use Context Managers?

Use them when:

* A resource needs to be **opened → used → closed**
* Cleanup is required even if errors occur
* You want concise, readable, safe code

Examples:

* File handling
* Networking
* Sessions
* Database transactions
* Locking/unlocking
* Timers
* Logging contexts

---

# ⭐ 9. Advantages of Context Managers

✔ Prevent resource leaks
✔ Error-safe cleanup
✔ Simplify code
✔ Highly Pythonic
✔ Improve readability

# **4. Pythonic Coding Patterns (Write Code Like a True Python Developer)**
----

Python has a unique programming style — clean, readable, expressive.
“Pythonic code” means writing code that follows these principles.

---

# 🧠 **1. Use List Comprehensions Instead of Loops**

### ❌ Non-Pythonic

```python
squares = []
for i in range(10):
    squares.append(i * i)
```

### ✔ Pythonic

```python
squares = [i * i for i in range(10)]
```

### Benefits:

* Shorter
* Faster
* More readable

---

# 🧠 **2. Use `enumerate()` Instead of `range(len())`**

### ❌ Non-Pythonic

```python
for i in range(len(data)):
    print(i, data[i])
```

### ✔ Pythonic

```python
for index, item in enumerate(data):
    print(index, item)
```

---

# 🧠 **3. Use `zip()` to Loop Over Multiple Iterables**

### ❌ Non-Pythonic

```python
for i in range(len(names)):
    print(names[i], ages[i])
```

### ✔ Pythonic

```python
for name, age in zip(names, ages):
    print(name, age)
```

---

# 🧠 **4. Use Unpacking for Clean Variable Assignments**

### Example:

```python
a, b, c = [1, 2, 3]
```

### Swap variables without temp:

```python
a, b = b, a
```

---

# 🧠 **5. Use `dict.get()` or `defaultdict`**

### ❌ Non-Pythonic

```python
if key in data:
    value = data[key]
else:
    value = 0
```

### ✔ Pythonic

```python
value = data.get(key, 0)
```

---

# 🧠 **6. Use `in` for Membership Checking**

### ❌ Non-Pythonic

```python
found = False
for item in items:
    if item == target:
        found = True
```

### ✔ Pythonic

```python
found = target in items
```

---

# 🧠 **7. Use F-strings for String Formatting**

### ❌ Non-Pythonic

```python
print("Name: %s, Age: %d" % (name, age))
```

### ✔ Pythonic

```python
print(f"Name: {name}, Age: {age}")
```

# 🧠 **8. Use Context Managers (`with`)**

Automatic resource cleanup.

### ✔ Pythonic

```python
with open("file.txt") as f:
    data = f.read()
```

---

# 🧠 **9. Use `try/except` Instead of Checking First (EAFP)**

Pythonic style prefers:

### **EAFP – Easier to Ask Forgiveness than Permission**

### ✔ Pythonic

```python
try:
    value = d[key]
except KeyError:
    value = 0
```

Rather than:

### ❌ LBYL – Look Before You Leap

```python
if key in d:
    value = d[key]
else:
    value = 0
```

# **5. Regular Expressions (RegEx) in Python**

Regular expressions allow you to **search**, **match**, and **manipulate** text using *patterns* instead of fixed characters.

Python provides the `re` module for working with regex.

```python
import re
```

---

# 🔥 1. What is a Regular Expression?

A **pattern-matching language** used to:

* validate data
* extract specific information
* replace text
* search text

Example pattern:
`r"\d{3}-\d{3}-\d{4}"`
→ matches a phone number like **123-456-7890**

---

# 🔥 2. Raw Strings `r" "` in Python

Always use **raw strings** for regex to avoid escaping issues.

### Example:

```python
pattern = r"\n"   # correct
```

Without raw string, Python thinks `\n` is a newline.

---

# 🔥 3. Basic Metacharacters

| Symbol | Meaning                      | Example              |      |      |
| ------ | ---------------------------- | -------------------- | ---- | ---- |
| `.`    | Any character except newline | `a.c` → "abc", "a-c" |      |      |
| `^`    | Start of string              | `^Hello`             |      |      |
| `$`    | End of string                | `world$`             |      |      |
| `*`    | 0 or more repetitions        | `ab*`                |      |      |
| `+`    | 1 or more repetitions        | `ab+`                |      |      |
| `?`    | 0 or 1 repetition            | `colou?r`            |      |      |
| `[]`   | Character set                | `[aeiou]`            |      |      |
| `[^]`  | Negation                     | `[^0-9]`             |      |      |
| `      | `                            | OR                   | `cat | dog` |
| `()`   | Grouping                     | `(ab)+`              |      |      |

---

# 🔥 4. Special Sequences

| Pattern | Meaning                      |
| ------- | ---------------------------- |
| `\d`    | Digit (0–9)                  |
| `\D`    | Non-digit                    |
| `\w`    | Word char (a-z, A-Z, 0-9, _) |
| `\W`    | Non-word                     |
| `\s`    | Whitespace                   |
| `\S`    | Non-whitespace               |

---

# 🔥 5. Quantifiers

| Pattern | Meaning               |
| ------- | --------------------- |
| `{n}`   | Exactly n times       |
| `{n,}`  | At least n times      |
| `{n,m}` | Between n and m times |

Example:
`r"\d{3}"` → matches **123**
`r"[A-Za-z]{2,4}"` → 2 to 4 letters

# 🔥 7. Grouping & Capturing

### Example:

```python
match = re.search(r"(\d{2})-(\d{2})-(\d{4})", "Date: 12-09-2025")
print(match.group(1))  # 12
print(match.group(2))  # 09
print(match.group(3))  # 2025
```

Groups allow extracting specific parts.

---

# 🔥 8. Named Groups

```python
pattern = r"(?P<day>\d{2})-(?P<month>\d{2})-(?P<year>\d{4})"
match = re.search(pattern, "12-09-2025")
print(match.group("year"))
```
