# **1. Virtual Environments & Project Structure**
-----

(venv, Project Organization, Requirements Files)

---

# 🔥 PART 1: **Virtual Environments (venv)**

A **virtual environment** is an isolated Python environment where you install libraries **specific to one project**, without affecting your system Python or other projects.

This prevents:

* Version conflicts
* Breaking old projects
* Global installations clutter

---

# ⭐ 1. What is `venv`?

`venv` = Virtual Environment module (built-in in Python 3)

It allows you to create a project-specific Python environment.

---

# ⭐ 2. Create a Virtual Environment

Navigate to your project folder:

```
cd myproject
```

### ✔ Create environment

```
python -m venv venv
```

This creates a folder named **venv** containing:

* isolated Python
* pip
* site-packages

---

# ⭐ 3. Activate Virtual Environment

### 🔹 On Windows:

```
venv\Scripts\activate
```

### 🔹 On Mac/Linux:

```
source venv/bin/activate
```

You’ll see:

```
(venv) C:\myproject>
```

Means it's active.

---

# ⭐ 4. Deactivate

```
deactivate
```

---

# ⭐ 5. Install Packages in venv

```
pip install requests pandas numpy
```

Only installed inside this environment — not global.

---

# ⭐ 6. Check installed packages

```
pip list
```

---

# ⭐ 7. Remove a package

```
pip uninstall package_name
```

---

# ⭐ 8. Delete virtual environment

Just delete the **venv/** folder.
It’s safe — everything is inside it.

---

# 🔥 PART 2: **Organizing Real Python Projects**

Here is a **professional project structure**:

```
myproject/
│
├── venv/               # virtual environment
├── requirements.txt    # dependencies
├── README.md           # project description
├── .gitignore          # ignore venv, cache files
├── src/                # main project folder
│   ├── __init__.py
│   ├── main.py
│   ├── utils.py
│   └── modules/
│       └── feature.py
│
├── tests/              # testing folder
│   └── test_main.py
│
└── data/               # datasets, input files
```

### ✔ Why this structure is good?

* Separation of code, tests, data
* Easy for others to understand
* Works well with Git, CI/CD
* Industry-standard layout

---

# ⭐ Folder Explanations

### ✔ `src/`

Your actual Python code lives here.

### ✔ `tests/`

Contains test files (pytest recommended).

### ✔ `data/`

CSV files, datasets, configs.

### ✔ `requirements.txt`

Stores libraries used by project.

### ✔ `README.md`

Explains how to run the project.

### ✔ `.gitignore`

Tells Git what **not to track**.

Typical `.gitignore` for Python:

```
venv/
__pycache__/
*.pyc
.env
```

---

# 🔥 PART 3: **Requirements Files**

A **requirements file** is a list of project dependencies (libraries).

### Create it:

```
pip freeze > requirements.txt
```

Example:

```
requests==2.31.0
pandas==2.0.3
numpy==1.25.0
```

---

# ⭐ Install dependencies from requirements:

```
pip install -r requirements.txt
```

This is how teams share project dependencies.

---

# ⭐ Key Commands (Quick Sheet)

| Task                 | Command                           |
| -------------------- | --------------------------------- |
| Create venv          | `python -m venv venv`             |
| Activate (Windows)   | `venv\Scripts\activate`           |
| Activate (Mac/Linux) | `source venv/bin/activate`        |
| Deactivate           | `deactivate`                      |
| Save requirements    | `pip freeze > requirements.txt`   |
| Install from file    | `pip install -r requirements.txt` |

---

# 🔥 Bonus: Advanced Project Structure (Pro-Level)

For larger projects:

```
myproject/
│
├── venv/
├── requirements.txt
├── setup.py            # for packaging
├── src/
│   └── myproject/
│       ├── __init__.py
│       ├── main.py
│       └── components/
├── docs/               # documentation
├── examples/           # example scripts
├── configs/            # config files
└── tests/
```

# **2. Version Control (Git Basics)**
----

Git is a **version control system** that helps you track changes in code, collaborate, experiment safely, and maintain clean project history.

---

# 🔥 1. Why Use Git?

✔ Save different versions of your project
✔ Undo mistakes
✔ Work with teams
✔ Create separate development branches
✔ Push your code to GitHub for backup and collaboration

---

# ⭐ 2. Git Installation Check

Check if Git is installed:

```
git --version
```

If not installed, install from:
**[https://git-scm.com/downloads](https://git-scm.com/downloads)**

---

# ⭐ 3. Configure Git (first-time only)

```
git config --global user.name "Your Name"
git config --global user.email "your_email@example.com"
```

---

# ⭐ 4. Initialize a Git Repository

Open your project folder and run:

```
git init
```

This creates a hidden folder `.git/` — it tracks all changes.

---

# ⭐ 5. Check Repository Status

```
git status
```

Shows:

* Modified files
* New files
* Untracked files
* Staged files

---

# ⭐ 6. Add Files to Staging Area

Add a single file:

```
git add filename.py
```

Add everything:

```
git add .
```

---

# ⭐ 7. Commit Your Changes

A commit = snapshot of your code.

```
git commit -m "Initial commit"
```

Commit messages should be meaningful.

---

# 🔥 8. Connecting to GitHub (Remote Repo)

### Step 1: Create a new repository on GitHub

(e.g., `myproject`)

### Step 2: Connect your local repo to GitHub

```
git remote add origin https://github.com/username/myproject.git
```

---

# ⭐ 9. Push Code to GitHub

First push (set upstream):

```
git push -u origin main
```

or if default branch is *master*:

```
git push -u origin master
```

Next pushes:

```
git push
```

---

# ⭐ 10. Pull Latest Changes

To update local repo with remote changes:

```
git pull
```

or:

```
git pull origin main
```

---

# 🧠 11. Branching Basics

Branches let you work on features without touching the main code.

### ✔ Create a branch

```
git branch feature-login
```

### ✔ Switch to that branch

```
git checkout feature-login
```

OR (modern way):

```
git switch feature-login
```

### ✔ Create + switch in one step

```
git checkout -b feature-login
```

---

# ⭐ 12. Merge Branch into Main

Switch to main:

```
git checkout main
```

Merge:

```
git merge feature-login
```

---

# ⭐ 13. Delete a Branch

After merging:

```
git branch -d feature-login
```

---

# ⭐ 14. View All Branches

```
git branch
```

---

# ⭐ 15. View Commit History

```
git log
```

Short one-line view:

```
git log --oneline
```

---

# 🎯 Git Workflow Summary

```
git init
git add .
git commit -m "message"
git remote add origin <repo-url>
git push -u origin main
```

Branch workflow:

```
git checkout -b new-feature
# work...
git add .
git commit -m "done feature"
git checkout main
git merge new-feature
git branch -d new-feature
```
