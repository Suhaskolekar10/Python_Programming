```python
# Reading data from file

file = open("data.txt", "r")
content = file.read()
print(content)

file.close()
```

```python
## Writing text to the file

file = open("data.txt", "w")
content = "I am writing file."

file.write(content)
file.close()
```

```python
file = open("data.txt", "r")
for line in file:
    print(line)

file.close()
```

```python
file = open("data.txt", "r")
line = file.readlines()

print(line)

file.close()
```

```python
# Using with statements  ---> Automatically close no needto close manualy.

with open("data.txt", "r") as file:
    content = file.read()

print(content)
```
