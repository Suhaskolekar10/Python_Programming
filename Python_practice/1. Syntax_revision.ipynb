### 3. Operators in Python

```python
# 1.Arithmatic operator.. (+,-,*,/,//,%,**)

a = 10
b = 3
print(f"{a} + {b} = {a + b}")
print(f"{a} - {b} = {a - b}")
print(f"{a} * {b} = {a * b}")
print(f"{a} / {b} = {a / b}")
print(f"{a} // {b} = {a // b}")
print(f"{a} % {b} = {a % b}")
print(f"{a} ** {b} = {a ** b}")
```

```python
# 2. Relational operators (==, !=, >, <, >=, <=) -->Result True/False

a = 25
b = 55

print(f"{a} == {b} = {a == b}")
print(f"{a} != {b} = {a != b}")
print(f"{a} > {b} = {a > b}")
print(f"{a} < {b} = {a < b}")
print(f"{a} >= {b} = {a >= b}")
print(f"{a} <= {b} = {a <= b}")
```

```python
# 3. Assignment operators (=, +=, -=, *=, /=, etc.) Used to assign values

a = 25

print(a)
a+=5
print(a)
a -= 5
print(a)
a *= 5
print(a)
a /= 5
print(a)
```

```python
# 4. Logical operator (Used to combine 2 or more conditions)

age = 25
is_graduate = True
is_married =False

# and     True --> When both conditions are true  
if age > 18 and is_graduate:
    print("Eligible for the job.")

# or        True ---> If any one condition is true
if age > 18 or is_graduate:
    print("Eligible to vote.")

# not          --- > Negate the current condition
if not is_married:
    print("Get married soon")
```

```python
#  5. Bitwise operator (&, |, ^, ~, >>, <<)

a = 5
b = 6

print(f"{a} & {b} = {a & b}")
print(f"{a} | {b} = {a & b}")
print(f"{a} ^ {b} = {a & b}")
print(f"~{a} = {~a}")
print(f"{a} >> 2  = {a >> 2}")
print(f"{a} << 2  = {a << 2}")
```

```python
# 6. Membership Operators (in , not in)

numbers = [1,2,3,4,5]

print(5 in numbers)

print(3 not in numbers)
```

```python
# 7. Identity operators (Check for memory reference not values..)

a = [1,2,3]
b = [1,2,3]
c = a

print(a is b)   # Assigning new values make different object... 

print(a is c)   # Assigning variable makes reference sharing...
```

### 4. Conditional statements

```python
# 1. if statement  --> We use this when we want to validate only one condition

user_age = int(input("Enter your age :- "))

if user_age > 18:
    print("You can vote.")
```

```python
# 2. if-else statement ---> when we have 2 conditions either first or second.

user_age = int(input("Enter your age."))

if user_age > 18:
    print("You can vote.")
else:
    print("You can not vote.")
```

```python
# 3. if-elif-else   ---> we use this when we have multiple condition to validate..

marks = int(input("Enter student's mark (0-100) : "))

if marks >= 85:
    print("Grade :- A")
elif marks >= 70:
    print("Grade :- B")
elif marks >= 55:
    print("Grade :- C")
elif marks >= 40:
    print("Grade :- D")
else:
    print("Grade :- Fail")
```

### 5. Loops in Python..

```python
# 1. for Loop  ---> Used when we know number of iterations...  (used over sequences)

# 1. Iterating over string

name = "Suhas"
for char in name:
    print(char)

print("--------------------------------------")

# 2. Iterating over list

numbers = [10,20,30,40,50,60]
for number in numbers:
    print(number)

print("--------------------------------------")

# 3. Iterating over tuple

values = (5,3,6,9,4)
for val in values:
    print(val)

print("--------------------------------------")

# 4. Iterating over range

for i in range(1,6):
    print(i)
```

```python
# 2. While loop  ---> Used when we don't know about no. of iteration (Condition based..)

# Countdown code

count = int(input("Enter value from where you want to start countdown :- "))
while count > 0:
    print(count)
    count -= 1
```

```python
# Loop control statements in Python

# 1) break  ---> it will break loop completely. No further operations..

for i in range(1,10):
    if i == 5:
        break
    print(i)
```

```python
# 2. Continue  ---> It will skips current iteration and move to the next 

for i in range(1,10):
    if i == 3:
        continue
    print(i)                     ## It will not print 3
```

```python
#  3. pass  ---> Do nothing, Acts as a placeholder for future use.. (It prevent loop breaking)

for i in range(10):
    pass
```

### 6. Functions in Python..

```python
# Function is a block of code written to do specific task..

# 1. Defining function

def greet():
    print("Welcome..")
```

```python
# 2. Calling a function

greet()
```

```python
# 3. Passing arguments to the function..

def addition(a, b):
    print(f"{a} + {b} = {a+b}")

addition(15,27)
```

```python
## 6. Variable Key-word Arguments

def info(**candidate_info):
    for key, value in candidate_info.items():
        print(f"{key} --> {value}")

info(Name = "Suhas", Age = 25, Height = 1.68)
```
