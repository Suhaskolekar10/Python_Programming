## **1. History of Python**
---

### **1. Creator**

* Python was created by **Guido van Rossum** in **1989** in the Netherlands.
* He wanted to create a language that was **easy to read, write, and maintain**.
* He named it **Python** after the British comedy group **Monty Python**, not the snake 🐍.

---

### **2. First Release**

* **Python 1.0** was released in **1991**.
* Features included:

  * Exception handling
  * Functions
  * Core data types like **str, list, dict**

---

### **3. Python 2.x Series**

* Released in **2000**, Python 2 introduced features like:

  * List comprehensions
  * Garbage collection
  * Unicode support (later versions)
* **Python 2.7** (released in 2010) was the last in the 2.x series.
* **End of Life:** January 1, 2020 – Python 2 is officially discontinued.

---

### **4. Python 3.x Series**

* **Python 3.0** was released in **2008** to fix design flaws in Python 2.
* Not backward compatible with Python 2.
* Improvements include:

  * Cleaner syntax for `print()` function
  * Better Unicode support
  * New libraries and standard library improvements
* Python 3 is **the current and actively maintained version**.

---

### **5. Key Features That Made Python Popular**

* **Easy to learn and readable syntax** – like English
* **Interpreted language** – code runs line by line
* **Open source and free**
* **Extensive libraries** – for math, AI, web, data science, etc.
* **Cross-platform** – works on Windows, Mac, Linux

---

### **6. Why Python is So Popular Today**

* Widely used in **AI, Data Science, Web Development, Automation, and Scripting**
* Huge **community support**
* Easy integration with other languages like C, C++, Java
* Thousands of **third-party libraries**

# **2. Basic Programming Terminologies**
---

## **1. Program**

* A **program** is a set of instructions that a computer executes to perform a task.
* Example: A program to calculate the sum of two numbers.

---

## **2. Programming Language**

* A **programming language** is a language we use to write programs.
* Examples: Python, Java, C++, JavaScript
* Python is **high-level**, meaning it’s closer to human language and easier to read.

---

## **3. Syntax**

* **Syntax** is the set of rules that defines how code must be written in a programming language.
* If you break syntax rules, the program will give an **error**.
* Example in Python: `print("Hello")` is correct, but `print "Hello"` is wrong in Python 3.

---

## **4. Statement**

* A **statement** is a single line of code that performs an action.
* Example: `x = 5` is a statement that assigns the value 5 to `x`.

---

## **5. Expression**

* An **expression** is a combination of values, variables, and operators that **evaluates to a single value**.
* Example: `2 + 3` evaluates to `5`.
* Example with variable: `x + 10` (if `x = 5`, the expression evaluates to `15`)

---

## **6. Variable**

* A **variable** is a name that stores a value in memory.
* Example:

  ```python
  age = 20
  name = "Suhas"
  ```
* `age` and `name` are variables storing data.

---

## **7. Data Types**

* **Data types** define the kind of data a variable can store.
* Common Python data types:

  * `int` → integer numbers (`5`, `100`)
  * `float` → decimal numbers (`3.14`)
  * `str` → text (`"Hello"`)
  * `bool` → True/False values

---

## **8. Operator**

* **Operators** perform operations on variables and values.
* Types:

  * Arithmetic: `+ - * / % ** //`
  * Comparison: `== != > < >= <=`
  * Logical: `and, or, not`

---

## **9. Function**

* A **function** is a reusable block of code that performs a specific task.
* Example: `print()` is a built-in function that displays output on the screen.
* You can also create your own functions with `def`.

---

## **10. Loop**

* A **loop** repeats a block of code multiple times.
* Two main types:

  * `for` loop → repeats for a fixed number of times
  * `while` loop → repeats until a condition is false

---

## **11. Conditional Statement**

* Conditional statements allow a program to **make decisions**.
* Example: `if`, `elif`, `else`
* Example: Check if a number is positive or negative.

---

## **12. Comment**

* A **comment** is a line in code ignored by the computer.
* Used to explain code for humans.
* In Python:

  ```python
  # This is a comment
  ''' This is a
  multi-line comment '''
  ```

---

## **13. Error**

* An **error** occurs when the program cannot run properly.
* Types:

  * **Syntax error** → code breaks the rules
  * **Runtime error** → error occurs while the program is running
  * **Logical error** → code runs but gives wrong result

# **3. Python’s basic syntax**

---

## **1. What is Syntax in Python?**

* **Syntax** is the set of rules that tells Python **how code should be written**.
* If you break syntax rules, Python will give a **SyntaxError**.

---

## **2. Key Rules of Python Syntax**

### **a) Case-Sensitivity**

* Python is **case-sensitive**.

  ```python
  name = "Suhas"
  Name = "John"
  print(name)  # Suhas
  print(Name)  # John
  ```
* `name` and `Name` are **two different variables**.

---

### **b) Indentation**

* Python uses **indentation (spaces or tabs)** to define blocks of code.
* No `{ }` like in C++ or Java.
* Standard: **4 spaces per indentation level**.

  ```python
  if 5 > 2:
      print("5 is greater than 2")
  ```

---

### **c) Comments**

* **Single-line comment:** starts with `#`
* **Multi-line comment:** inside `''' '''` or `""" """`

  ```python
  # This is a single-line comment
  '''
  This is a
  multi-line comment
  '''
  ```

---

### **d) Statements**

* A **statement** is a single line of code that does something.
* Example:

  ```python
  x = 5
  print(x)
  ```
* You can write multiple statements on **one line** using `;` (not recommended):

  ```python
  x = 5; y = 10; print(x + y)
  ```

---

### **e) Python is Interpreted**

* Python executes code **line by line**, so errors are easier to find.
* You can run Python:

  * **Interactively**: using `python` in terminal
  * **As a script**: saving code in `.py` file and running it

---

### **f) Python is Case-Sensitive**

* Python treats uppercase and lowercase letters as different:

  ```python
  Age = 20
  age = 25
  print(Age)  # 20
  print(age)  # 25
  ```

---

### **g) Python Statements can Span Multiple Lines**

* Using **backslash `\`**:

  ```python
  total = 1 + 2 + 3 + \
          4 + 5
  print(total)  # 15
  ```
* Or inside brackets `()`, `{}`, `[]`:

  ```python
  numbers = [1, 2, 3,
             4, 5, 6]
  print(numbers)
  ```

---

### **h) Python is Dynamically Typed**

* You **don’t need to declare the type** of a variable.

  ```python
  x = 5        # integer
  x = "Hello"  # string
  ```

---

### **i) Printing Output**

* Use `print()` to display output.

  ```python
  print("Hello, Python!")
  print(2 + 3)
  name = "Suhas"
  print("My name is", name)
  ```

---

### **j) Taking Input**

* Use `input()` function to get user input (always returns string):

  ```python
  name = input("Enter your name: ")
  print("Hello,", name)
  age = int(input("Enter your age: "))  # convert string to int
  print("You are", age, "years old")
  ```
